## 在链式调用中使用

You can use `use` in a chain to apply multiple middleware functions to a request.

您可以在链式调用中使用 `use` 来将多个中间件函数应用于请求。

```javascript
import { Hono } from 'hono'

const app = new Hono()

const logger = async (c, next) => {
  console.log(`${c.req.method} ${c.req.url}`)
  await next()
}

const auth = async (c, next) => {
  if (c.req.header('Authorization')) {
    await next()
  } else {
    return c.text('Unauthorized', 401)
  }
}

app.use('/api/*', logger, auth) // Apply logger and auth middleware to /api/* routes

app.get('/api/users', (c) => {
  return c.text('User list')
})

app.get('/api/posts', (c) => {
  return c.text('Post list')
})

app.get('/', (c) => {
  return c.text('Hello Hono!')
})

export default app
```

In this example, both `logger` and `auth` middleware functions are applied to any request that matches the `/api/*` path. If the `Authorization` header is not present, the request will be stopped at the `auth` middleware and a 401 Unauthorized response will be returned.

在此示例中，`logger` 和 `auth` 中间件函数都应用于匹配 `/api/*` 路径的任何请求。如果不存在 `Authorization` 头，请求将在 `auth` 中间件处停止，并返回 401 Unauthorized 响应。

You can also chain `use` calls to apply middleware to specific paths.

您还可以链式调用 `use` 来将中间件应用于特定路径。

```javascript
import { Hono } from 'hono'

const app = new Hono()

const logger = async (c, next) => {
  console.log(`${c.req.method} ${c.req.url}`)
  await next()
}

const adminOnly = async (c, next) => {
  // Check if the user is an admin
  if (c.req.header('X-Admin')) {
    await next()
  } else {
    return c.text('Forbidden', 403)
  }
}

app.use('/admin/*', logger) // Apply logger middleware to all /admin/* routes
app.use('/admin/*', adminOnly) // Apply adminOnly middleware to all /admin/* routes

app.get('/admin/dashboard', (c) => {
  return c.text('Admin Dashboard')
})

export default app
```

In this example, the `logger` middleware is applied first to all `/admin/*` routes, followed by the `adminOnly` middleware. This means that all requests to `/admin/*` will be logged, and then checked for admin privileges.

在此示例中，`logger` 中间件首先应用于所有 `/admin/*` 路由，然后是 `adminOnly` 中间件。这意味着所有对 `/admin/*` 的请求都将被记录，然后检查管理员权限。

**Note:** Middleware functions are executed in the order they are registered.

**注意：** 中间件函数按注册顺序执行。
