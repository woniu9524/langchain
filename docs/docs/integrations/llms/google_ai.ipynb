## 在链中使用

You can use `use` to compose multiple hooks together. This is useful for creating reusable logic.

您可以使用 `use` 将多个钩子组合在一起。这对于创建可重用逻辑非常有用。

```javascript
import { use, useEffect } from 'react';

function useFetch(url) {
  const [data, setData] = useState(null);
  const [loading, setLoading] = useState(true);
  const [error, setError] = useState(null);

  useEffect(() => {
    setLoading(true);
    fetch(url)
      .then(res => res.json())
      .then(data => {
        setData(data);
        setLoading(false);
      })
      .catch(error => {
        setError(error);
        setLoading(false);
      });
  }, [url]);

  return { data, loading, error };
}

function useUserData(userId) {
  const { data, loading, error } = useFetch(`/api/users/${userId}`);
  return { user: data, loading, error };
}

function UserProfile({ userId }) {
  const { user, loading, error } = useUserData(userId);

  if (loading) return <p>Loading...</p>;
  if (error) return <p>Error: {error.message}</p>;

  return (
    <div>
      <h1>{user.name}</h1>
      <p>{user.email}</p>
    </div>
  );
}
```

In the example above, `useUserData` uses `useFetch` to fetch user data. This allows us to compose reusable fetching logic.

在上面的示例中，`useUserData` 使用 `useFetch` 来获取用户数据。这使我们能够组合可重用的获取逻辑。

You can also use `use` to pass data between hooks. For example, you could have a hook that fetches data and another hook that processes it.

您还可以使用 `use` 在钩子之间传递数据。例如，您可以有一个获取数据的钩子，以及另一个处理该数据的钩子。

```javascript
import { use, useEffect, useState } from 'react';

function useFetch(url) {
  const [data, setData] = useState(null);
  const [loading, setLoading] = useState(true);
  const [error, setError] = useState(null);

  useEffect(() => {
    setLoading(true);
    fetch(url)
      .then(res => res.json())
      .then(data => {
        setData(data);
        setLoading(false);
      })
      .catch(error => {
        setError(error);
        setLoading(false);
      });
  }, [url]);

  return { data, loading, error };
}

function useProcessedData(data) {
  const [processedData, setProcessedData] = useState(null);

  useEffect(() => {
    if (data) {
      // Process the data here
      setProcessedData(data.map(item => item.toUpperCase()));
    }
  }, [data]);

  return { processedData };
}

function MyComponent({ url }) {
  const { data, loading, error } = useFetch(url);
  const { processedData } = useProcessedData(data);

  if (loading) return <p>Loading...</p>;
  if (error) return <p>Error: {error.message}</p>;

  return (
    <ul>
      {processedData && processedData.map(item => <li key={item}>{item}</li>)}
    </ul>
  );
}
```

In this example, `useProcessedData` uses the data returned by `useFetch` to process it. This allows us to create more complex logic by composing simpler hooks.

在此示例中，`useProcessedData` 使用 `useFetch` 返回的数据对其进行处理。这使我们能够通过组合更简单的钩子来创建更复杂的逻辑。

## 流式通话

You can stream audio from a call to your application.
您可以将通话音频流式传输到您的应用程序。

This feature is available for both incoming and outgoing calls.
此功能可用于来电和去电。

### Requirements
### 要求

- You need to have the `audio` capability enabled for your app.
- 您需要在您的应用中启用 `audio` 功能。
- You need to have the `incomingCall` capability enabled for your app if you want to stream audio from incoming calls.
- 如果您想从传入电话流式传输音频，则需要为您的应用启用 `incomingCall` 功能。

### Streaming audio from incoming calls
### 从传入电话流式传输音频

To stream audio from an incoming call, you need to:
要从传入电话流式传输音频，您需要：

1.  **Accept the call.**
    1.  **接听电话。**
2.  **Start streaming the audio.**
    2.  **开始流式传输音频。**

```javascript
// Example of accepting a call and streaming audio
// 接受电话并流式传输音频的示例

import { CallClient, CallAgent } from "@azure/communication-calling";
import { AzureCommunicationTokenCredential } from "@azure/communication-common";

const token = "<YOUR_TOKEN>";
const endpointUrl = "<YOUR_ENDPOINT_URL>";
const groupId = "<YOUR_GROUP_ID>"; // Optional: For group calls

const credential = new AzureCommunicationTokenCredential(token);
const callClient = new CallClient();
const callAgent = await callClient.createCallAgent(credential, {
    displayName: "Your Name"
});

// Listen for incoming calls
// 监听传入电话
callAgent.on("incomingCall", async (callEventArgs) => {
    const call = callEventArgs.call;

    // Accept the call
    // 接听电话
    await call.accept();

    // Start streaming audio
    // 开始流式传输音频
    const stream = await call.startStreamingAudio();

    // You can now process the audio stream
    // 您现在可以处理音频流
    stream.on("data", (data) => {
        console.log("Received audio data:", data);
    });
});
```

### Streaming audio from outgoing calls
### 从去电流式传输音频

To stream audio from an outgoing call, you need to:
要从去电流式传输音频，您需要：

1.  **Start the call.**
    1.  **发起电话。**
2.  **Start streaming the audio.**
    2.  **开始流式传输音频。**

```javascript
// Example of starting a call and streaming audio
// 发起电话并流式传输音频的示例

import { CallClient, CallAgent } from "@azure/communication-calling";
import { AzureCommunicationTokenCredential } from "@azure/communication-common";

const token = "<YOUR_TOKEN>";
const endpointUrl = "<YOUR_ENDPOINT_URL>";
const calleeId = "<CALLEE_ID>"; // The ID of the person you want to call

const credential = new AzureCommunicationTokenCredential(token);
const callClient = new CallClient();
const callAgent = await callClient.createCallAgent(credential, {
    displayName: "Your Name"
});

// Start the call
// 发起电话
const call = await callAgent.startCall([{ id: calleeId }], {});

// Start streaming audio
// 开始流式传输音频
const stream = await call.startStreamingAudio();

// You can now process the audio stream
// 您现在可以处理音频流
stream.on("data", (data) => {
    console.log("Received audio data:", data);
});
```

### Stopping audio streaming
### 停止音频流式传输

To stop streaming audio, you can call the `stopStreamingAudio` method on the `Call` object.
要停止音频流式传输，您可以调用 `Call` 对象上的 `stopStreamingAudio` 方法。

```javascript
// Example of stopping audio streaming
// 停止音频流式传输的示例

await call.stopStreamingAudio();

### 安全设置

Gemini 模型具有默认的安全设置，可以被覆盖。如果您从模型中收到大量“安全警告”，可以尝试调整模型的 `safety_settings` 属性。例如，要关闭危险内容的安全性阻止，您可以按以下方式构建您的 LLM：

````python
from langchain_google_genai import GoogleGenerativeAI, HarmBlockThreshold, HarmCategory

llm = GoogleGenerativeAI(
    model="gemini-pro",
    google_api_key=api_key,
    safety_settings={
        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE,
    },
)
````

如需查看可用类别和阈值的枚举，请参阅 Google 的 [安全设置类型](https://ai.google.dev/api/python/google/generativeai/types/SafetySettingDict)。
