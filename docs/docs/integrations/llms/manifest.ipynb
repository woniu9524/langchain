```python
from langchain_community.llms.manifest import ManifestWrapper
from manifest import Manifest
```

```python
manifest = Manifest(
    client_name="huggingface", client_connection="http://127.0.0.1:5000"
)
print(manifest.client_pool.get_current_client().get_model_params())
```

```python
llm = ManifestWrapper(
    client=manifest, llm_kwargs={"temperature": 0.001, "max_tokens": 256}
)
```

```python
# Map reduce example
from langchain.chains.mapreduce import MapReduceChain
from langchain_core.prompts import PromptTemplate
from langchain_text_splitters import CharacterTextSplitter

_prompt = """Write a concise summary of the following:


{text}


CONCISE SUMMARY:"""
prompt = PromptTemplate.from_template(_prompt)

text_splitter = CharacterTextSplitter()

mp_chain = MapReduceChain.from_params(llm, prompt, text_splitter)
```

```python
with open("../../how_to/state_of_the_union.txt") as f:
    state_of_the_union = f.read()
mp_chain.run(state_of_the_union)
```

## 比较 HF 模型

This document compares the performance of various Hugging Face (HF) models on a range of tasks.

本文档比较了各种 Hugging Face (HF) 模型在一系列任务上的性能。

### Models

### 模型

*   **BERT (Bidirectional Encoder Representations from Transformers)**: A foundational model for many NLP tasks.
*   **RoBERTa (A Robustly Optimized BERT Pretraining Approach)**: An optimized version of BERT.
*   **GPT-2 (Generative Pre-trained Transformer 2)**: A powerful generative model.
*   **T5 (Text-to-Text Transfer Transformer)**: A versatile model that frames all NLP tasks as text-to-text problems.
*   **DistilBERT**: A smaller, faster, and lighter version of BERT.

*   **BERT (Bidirectional Encoder Representations from Transformers)**：许多 NLP 任务的基础模型。
*   **RoBERTa (A Robustly Optimized BERT Pretraining Approach)**：BERT 的优化版本。
*   **GPT-2 (Generative Pre-trained Transformer 2)**：一个强大的生成模型。
*   **T5 (Text-to-Text Transfer Transformer)**：一个多功能模型，将所有 NLP 任务都视为文本到文本的问题。
*   **DistilBERT**：BERT 的更小、更快、更轻的版本。

### Tasks

### 任务

*   **Text Classification**: Assigning a category label to a given text.
*   **Named Entity Recognition (NER)**: Identifying and categorizing named entities in text.
*   **Question Answering**: Answering questions based on a given context.
*   **Summarization**: Generating a concise summary of a longer text.
*   **Translation**: Translating text from one language to another.

*   **文本分类**：为给定文本分配类别标签。
*   **命名实体识别 (NER)**：识别和分类文本中的命名实体。
*   **问答**：根据给定上下文回答问题。
*   **摘要**：为较长的文本生成简洁的摘要。
*   **翻译**：将文本从一种语言翻译成另一种语言。

### Conclusion

### 结论

The choice of model depends heavily on the specific task requirements and available computational resources. For tasks requiring high accuracy on standard NLP benchmarks, RoBERTa is often a strong contender. For a versatile model capable of handling diverse text-to-text tasks, T5 is an excellent choice. When efficiency is paramount, DistilBERT provides a compelling alternative. GPT-2 remains a top choice for pure text generation.

模型的选择在很大程度上取决于具体的任务需求和可用的计算资源。对于在标准 NLP 基准测试中需要高准确率的任务，RoBERTa 通常是一个强有力的竞争者。对于能够处理各种文本到文本任务的多功能模型，T5 是一个绝佳的选择。当效率至关重要时，DistilBERT 提供了一个引人注目的替代方案。GPT-2 在纯文本生成方面仍然是首选。

```python
from langchain.model_laboratory import ModelLaboratory

manifest1 = ManifestWrapper(
    client=Manifest(
        client_name="huggingface", client_connection="http://127.0.0.1:5000"
    ),
    llm_kwargs={"temperature": 0.01},
)
manifest2 = ManifestWrapper(
    client=Manifest(
        client_name="huggingface", client_connection="http://127.0.0.1:5001"
    ),
    llm_kwargs={"temperature": 0.01},
)
manifest3 = ManifestWrapper(
    client=Manifest(
        client_name="huggingface", client_connection="http://127.0.0.1:5002"
    ),
    llm_kwargs={"temperature": 0.01},
)
llms = [manifest1, manifest2, manifest3]
model_lab = ModelLaboratory(llms)
```
