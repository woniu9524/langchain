This guide will walk you through the process of creating a graph connection and inserting demo data into your graph database.

本指南将引导您完成创建图连接并将演示数据插入图数据库的过程。

**Prerequisites**

**先决条件**

*   A running instance of your graph database.
*   Access credentials (username, password, host, port) for your graph database.

*   正在运行的图数据库实例。
*   图数据库的访问凭证（用户名、密码、主机、端口）。

**Steps**

**步骤**

1.  **Open the Graph Studio:**
    *   Navigate to your graph database's web interface. This is typically accessed via a web browser.

        *   使用您提供的凭证登录。

2.  **Create a New Connection:**
    *   In Graph Studio, locate the option to create a new connection. This might be a button labeled "New Connection," "Add Database," or similar.
    *   A dialog or form will appear, prompting you for connection details.

    2.  **创建新连接：**
        *   在 Graph Studio 中，找到创建新连接的选项。这可能是一个标有“新建连接”、“添加数据库”或类似的按钮。
        *   将出现一个对话框或表单，提示您输入连接详细信息。

3.  **Enter Connection Details:**
    *   **Database Type:** Select the type of graph database you are using (e.g., Neo4j, ArangoDB, Amazon Neptune).
    *   **Host:** Enter the hostname or IP address of your graph database server.
    *   **Port:** Enter the port number your graph database is listening on.
    *   **Username:** Enter your database username.
    *   **Password:** Enter your database password.
    *   **Database Name (if applicable):** Some databases require a specific database name.

    3.  **输入连接详细信息：**
        *   **数据库类型：** 选择您正在使用的图数据库类型（例如，Neo4j、ArangoDB、Amazon Neptune）。
        *   **主机：** 输入图数据库服务器的主机名或 IP 地址。
        *   **端口：** 输入图数据库正在监听的端口号。
        *   **用户名：** 输入您的数据库用户名。
        *   **密码：** 输入您的数据库密码。
        *   **数据库名称（如果适用）：** 某些数据库需要特定的数据库名称。

4.  **Test the Connection:**
    *   Most Graph Studio tools provide a "Test Connection" button. Click this to verify that your credentials and connection details are correct.
    *   If the test fails, double-check all the entered information.

    4.  **测试连接：**
        *   大多数 Graph Studio 工具都提供“测试连接”按钮。单击此按钮以验证您的凭证和连接详细信息是否正确。

    *   Once the connection is successfully tested, click "Save" or "Connect" to establish the connection.

        *   成功测试连接后，单击“保存”或“连接”以建立连接。

        *   **Import from File:** Upload a file (e.g., CSV, JSON) containing your demo data.

**Example: Inserting Demo Data using Cypher (for Neo4j)**

**示例：使用 Cypher 插入演示数据（适用于 Neo4j）**

Let's assume you want to create a simple social network graph with `Person` nodes and `FRIENDS_WITH` relationships.

假设您想创建一个简单的社交网络图，包含 `Person` 节点和 `FRIENDS_WITH` 关系。

```cypher
// Create Person nodes
CREATE (:Person {name: 'Alice', age: 30});
CREATE (:Person {name: 'Bob', age: 25});
CREATE (:Person {name: 'Charlie', age: 35});

// Create relationships
MATCH (a:Person {name: 'Alice'}), (b:Person {name: 'Bob'})
CREATE (a)-[:FRIENDS_WITH]->(b);

MATCH (a:Person {name: 'Bob'}), (b:Person {name: 'Charlie'})
CREATE (a)-[:FRIENDS_WITH]->(b);

MATCH (a:Person {name: 'Alice'}), (b:Person {name: 'Charlie'})
CREATE (a)-[:FRIENDS_WITH]->(b);
```

**Explanation of the Cypher Query:**

**Cypher 查询说明：**

*   `CREATE (:Person {name: 'Alice', age: 30})` creates a node with the label `Person` and sets its `name` property to 'Alice' and `age` property to 30.
*   `MATCH (a:Person {name: 'Alice'}), (b:Person {name: 'Bob'})` finds the nodes representing Alice and Bob.
*   `CREATE (a)-[:FRIENDS_WITH]->(b)` creates an outgoing relationship of type `FRIENDS_WITH` from node `a` (Alice) to node `b` (Bob).

*   `CREATE (:Person {name: 'Alice', age: 30})` 创建一个标签为 `Person` 的节点，并将其 `name` 属性设置为 'Alice'，`age` 属性设置为 30。
*   `MATCH (a:Person {name: 'Alice'}), (b:Person {name: 'Bob'})` 查找代表 Alice 和 Bob 的节点。
*   `CREATE (a)-[:FRIENDS_WITH]->(b)` 在节点 `a`（Alice）和节点 `b`（Bob）之间创建了一个类型为 `FRIENDS_WITH` 的出站关系。

You can execute these queries directly in the Graph Studio's query editor.

您可以在 Graph Studio 的查询编辑器中直接执行这些查询。

**Verification**

**验证**

After inserting the data, you can verify it by:

插入数据后，您可以通过以下方式进行验证：

*   **Running a query to count nodes and relationships:**
    ```cypher
    MATCH (n) RETURN count(n);
    ```

## 创建 FalkorDBQAChain

FalkorDBQAChain 是一个用于与 FalkorDB 交互的链。它允许您使用自然语言问题查询数据库。

**先决条件**

*   **FalkorDB 实例：** 您需要一个正在运行的 FalkorDB 实例。
*   **安装 FalkorDB Python 客户端：**
    ```bash
    pip install falkordb
    ```

**使用 FalkorDBQAChain**

以下是如何使用 FalkorDBQAChain 的示例：

```python
from falkordb import FalkorDB
from langchain.chains import FalkorDBQAChain
from langchain.chat_models import ChatOpenAI

# 连接到 FalkorDB
db = FalkorDB(host="localhost", port=7890, database="falkordb")

# 初始化语言模型
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)

# 创建 FalkorDBQAChain
falkordb_qa_chain = FalkorDBQAChain.from_llm(
    llm=llm,
    db=db,
    k=10,  # 返回最相似的 10 个节点
    search_index="falkordb_index",  # 您的 FalkorDB 索引名称
    verbose=True,
)

# 提出问题
question = "Who are the main characters in the movie The Matrix?"
response = falkordb_qa_chain.run(question)

print(response)
```

**参数说明**

*   `llm`: 您选择的语言模型。
*   `db`: FalkorDB 实例。
*   `k`: 从 FalkorDB 中检索最相似的节点数量。
*   `search_index`: 您在 FalkorDB 中使用的搜索索引的名称。
*   `verbose`: 如果设置为 `True`，则会打印链的详细信息。

**工作原理**

1.  **问题解析：** FalkorDBQAChain 接收您的自然语言问题。
2.  **查询生成：** 它使用语言模型将您的问题转换为 FalkorDB 可以理解的 Cypher 查询。
3.  **数据库查询：** 生成的 Cypher 查询在您的 FalkorDB 实例上执行。
4.  **结果检索：** 查询结果（通常是节点和关系）被检索出来。
5.  **答案生成：** 语言模型接收查询结果，并根据这些结果生成一个自然语言答案。

**自定义**

您可以根据您的具体需求自定义 FalkorDBQAChain：

*   **自定义 Cypher 查询：** 您可以提供自定义的 Cypher 查询模板，以更精确地控制数据库交互。
*   **调整 `k` 值：** 实验不同的 `k` 值以找到最适合您用例的平衡点。
*   **集成其他 Langchain 组件：** 您可以将 FalkorDBQAChain 与 Langchain 中的其他链和代理结合使用，以创建更复杂的应用程序。

## 查询图谱

You can query the graph using GraphQL.

您可以使用 GraphQL 查询图谱。

```graphql
query GetUser($userId: ID!) {
  user(id: $userId) {
    id
    name
    email
    posts {
      id
      title
      body
    }
  }
}
```

```json
{
  "userId": "1"
}
```

The above query will fetch a user by their ID, along with their posts.

上面的查询将根据用户 ID 获取用户及其发布的帖子。

### Arguments

### 参数

Arguments are used to filter and specify the data you want to retrieve.

参数用于过滤和指定您想要检索的数据。

- `id` (ID!): The ID of the user to retrieve.

- `id` (ID!): 要检索的用户的 ID。

### Fields

### 字段

Fields are the specific pieces of data you want to get back from the server.

字段是从服务器获取的特定数据片段。

- `id` (ID!): The unique identifier for the user.

- `id` (ID!): 用户的唯一标识符。

- `name` (String!): The name of the user.

- `name` (String!): 用户的姓名。

- `email` (String!): The email address of the user.

- `email` (String!): 用户的电子邮件地址。

- `posts` ([Post!]!): A list of posts created by the user.

- `posts` ([Post!]!): 用户创建的帖子列表。

### Nested Fields

### 嵌套字段

You can also query nested fields, such as the `title` and `body` of a user's posts.

您还可以查询嵌套字段，例如用户帖子的 `title`（标题）和 `body`（正文）。

- `id` (ID!): The unique identifier for the post.

- `id` (ID!): 帖子的唯一标识符。

- `title` (String!): The title of the post.

- `title` (String!): 帖子的标题。

- `body` (String!): The body of the post.

- `body` (String!): 帖子的正文。
