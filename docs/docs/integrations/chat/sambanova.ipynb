## 实例化

现在我们可以实例化我们的模型对象并生成聊天补全：

````python
from langchain_sambanova import ChatSambaNovaCloud

llm = ChatSambaNovaCloud(
    model="Meta-Llama-3.3-70B-Instruct",
    max_tokens=1024,
    temperature=0.7,
    top_p=0.01,
)
````

## 调用

````python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to French. "
        "Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = llm.invoke(messages)
ai_msg
````

````python
print(ai_msg.content)
````

## 链式调用

我们可以像这样将模型与提示模板进行[链式调用](/docs/how_to/sequence/)：

````python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate(
    [
        (
            "system",
            "You are a helpful assistant that translates {input_language} "
            "to {output_language}.",
        ),
        ("human", "{input}"),
    ]
)

chain = prompt | llm
chain.invoke(
    {
        "input_language": "English",
        "output_language": "German",
        "input": "I love programming.",
    }
)
````

## 流式传输

````python
system = "You are a helpful assistant with pirate accent."
human = "I want to learn more about this animal: {animal}"
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", human)])

chain = prompt | llm

for chunk in chain.stream({"animal": "owl"}):
    print(chunk.content, end="", flush=True)
````

## 异步流

Async Streaming 是一种用于处理数据流的模式，它允许您在不阻塞主线程的情况下，异步地发送和接收数据。这对于处理大量数据或需要实时响应的应用程序非常有用。

### 核心概念

*   **流 (Stream):** 数据序列，可以随时间推移而生成。
*   **异步 (Async):** 操作不会立即完成，而是返回一个承诺（Promise），表示操作将在未来完成。
*   **生产者 (Producer):** 生成流中数据的实体。
*   **消费者 (Consumer):** 消费流中数据的实体。

### 工作原理

Async Streaming 的工作原理是，生产者异步地生成数据，并将数据“推送”给消费者。消费者可以订阅生产者，并在数据可用时接收数据。这个过程是非阻塞的，这意味着生产者和消费者可以同时执行其他任务，而不会互相等待。

### 常见用例

*   **实时数据处理:** 处理来自传感器、日志文件或网络套接字的数据。
*   **文件下载和上传:** 在下载或上传大文件时提供进度反馈，并允许用户在操作进行时继续使用应用程序。
*   **API 集成:** 从外部 API 获取大量数据，并逐步处理，而不是一次性加载所有数据。
*   **用户界面更新:** 当后台任务生成新数据时，异步更新用户界面，提供流畅的用户体验。

### 在 JavaScript 中实现 Async Streaming

JavaScript 提供了 `ReadableStream` 和 `WritableStream` 等 API 来支持异步流。

```javascript
// 示例：创建一个简单的异步流
const readableStream = new ReadableStream({
  start(controller) {
    let counter = 0;
    const intervalId = setInterval(() => {
      if (counter < 5) {
        const data = `数据 ${counter}`;
        controller.enqueue(data); // 将数据推送到流中
        console.log(`生产者: 发送 "${data}"`);
        counter++;
      } else {
        clearInterval(intervalId);
        controller.close(); // 关闭流
        console.log("生产者: 流已关闭");
      }
    }, 1000);
  }
});

// 示例：消费异步流
async function consumeStream(stream) {
  const reader = stream.getReader();
  while (true) {
    const { done, value } = await reader.read(); // 异步读取数据
    if (done) {
      console.log("消费者: 流已结束");
      break;
    }
    console.log(`消费者: 接收 "${value}"`);
  }
}

consumeStream(readableStream);
```

### 关键 API

*   **`ReadableStream`:** 表示可读的数据流。
    *   `getReader()`: 获取一个 `ReadableStreamDefaultReader` 来读取流中的数据。
    *   `enqueue(chunk)`: 将一个数据块添加到流中。
    *   `close()`: 关闭流，表示没有更多数据可用。
    *   `error(e)`: 使用错误关闭流。
*   **`WritableStream`:** 表示可写的数据流。
    *   `getWriter()`: 获取一个 `WritableStreamDefaultWriter` 来向流写入数据。
    *   `write(chunk)`: 将一个数据块写入流。
    *   `close()`: 关闭流。
    *   `abort(reason)`: 使用原因中止流。
*   **`ReadableStreamDefaultReader`:** 用于从 `ReadableStream` 读取数据块。
    *   `read()`: 返回一个包含 `done` 和 `value` 属性的 Promise。
*   **`WritableStreamDefaultWriter`:** 用于向 `WritableStream` 写入数据块。
    *   `write(chunk)`: 将数据块写入流。
    *   `close()`: 关闭流。
    *   `abort(reason)`: 中止流。

### 优势

*   **提高性能:** 通过异步处理，避免阻塞主线程，提高应用程序的响应能力。
*   **内存效率:** 允许处理大于可用内存的数据集，因为数据是分块处理的。
*   **更好的用户体验:** 提供实时反馈和流畅的交互。
*   **可组合性:** 流可以轻松地组合和转换，以实现复杂的数据处理管道。

### 挑战

*   **错误处理:** 需要仔细处理流中的错误，以确保应用程序的健壮性。
*   **背压 (Backpressure):** 当生产者生成数据的速度快于消费者处理数据的速度时，需要一种机制来控制生产者的速度，以防止内存溢出。`ReadableStream` 和 `WritableStream` API 内置了对背压的支持。
*   **理解复杂性:** 异步编程本身就具有一定的复杂性，异步流增加了另一层抽象。

Async Streaming 是现代 Web 开发中处理数据流的强大模式，理解其核心概念和 API 对于构建高性能、响应迅速的应用程序至关重要。

````python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "human",
            "in less than {num_words} words explain me {topic} ",
        )
    ]
)
chain = prompt | llm

async for chunk in chain.astream({"num_words": 30, "topic": "quantum computers"}):
    print(chunk.content, end="", flush=True)
````

## 工具调用

````python
from datetime import datetime

from langchain_core.messages import HumanMessage, ToolMessage
from langchain_core.tools import tool


@tool
def get_time(kind: str = "both") -> str:
    """Returns current date, current time or both.
    Args:
        kind(str): date, time or both
    Returns:
        str: current date, current time or both
    """
    if kind == "date":
        date = datetime.now().strftime("%m/%d/%Y")
        return f"Current date: {date}"
    elif kind == "time":
        time = datetime.now().strftime("%H:%M:%S")
        return f"Current time: {time}"
    else:
        date = datetime.now().strftime("%m/%d/%Y")
        time = datetime.now().strftime("%H:%M:%S")
        return f"Current date: {date}, Current time: {time}"


tools = [get_time]


def invoke_tools(tool_calls, messages):
    available_functions = {tool.name: tool for tool in tools}
    for tool_call in tool_calls:
        selected_tool = available_functions[tool_call["name"]]
        tool_output = selected_tool.invoke(tool_call["args"])
        print(f"Tool output: {tool_output}")
        messages.append(ToolMessage(tool_output, tool_call_id=tool_call["id"]))
    return messages
````

````python
llm_with_tools = llm.bind_tools(tools=tools)
messages = [
    HumanMessage(
        content="I need to schedule a meeting for two weeks from today. "
        "Can you tell me the exact date of the meeting?"
    )
]
````

````python
response = llm_with_tools.invoke(messages)
while len(response.tool_calls) > 0:
    print(f"Intermediate model response: {response.tool_calls}")
    messages.append(response)
    messages = invoke_tools(response.tool_calls, messages)
    response = llm_with_tools.invoke(messages)

print(f"final response: {response.content}")
````

## 结构化输出

This guide explains how to use structured outputs to get more predictable and reusable results from your LLM.

本指南将介绍如何使用结构化输出来从 LLM 获取更可预测和可重用的结果。

Structured outputs are a way to guide the LLM to return its response in a specific format, such as JSON or YAML. This makes it easier to parse and use the LLM's output in your applications.

结构化输出是一种引导 LLM 以特定格式（如 JSON 或 YAML）返回其响应的方式。这使得在您的应用程序中解析和使用 LLM 的输出更加容易。

### Why use structured outputs?

### 为什么使用结构化输出？

*   **Predictability:** LLMs can sometimes be unpredictable in their output format. Structured outputs ensure you get data in a consistent format every time.
*   **Reusability:** By having your LLM output data in a structured format, you can easily integrate it into your existing applications and workflows.
*   **Easier Parsing:** Machine-readable formats like JSON and YAML are much easier for applications to parse than free-form text.
*   **Reduced Errors:** Less manual parsing means fewer opportunities for errors.

*   **可预测性：** LLM 的输出格式有时可能难以预测。结构化输出可确保您每次都能获得一致格式的数据。
*   **可重用性：** 通过让 LLM 以结构化格式输出数据，您可以轻松地将其集成到现有的应用程序和工作流中。
*   **更轻松的解析：** JSON 和 YAML 等机器可读格式比自由格式文本更容易被应用程序解析。
*   **减少错误：** 更少的手动解析意味着更少出错的机会。

### How to use structured outputs

### 如何使用结构化输出

The most common way to use structured outputs is by providing a schema or template for the LLM to follow. Here are a few popular methods:

使用结构化输出最常见的方法是提供 LLM 需要遵循的模式或模板。以下是一些流行的方法：

1.  **JSON Schema:** You can define a JSON schema that describes the structure of the desired JSON output. The LLM will then try to generate JSON that conforms to this schema.

    1.  **JSON Schema：** 您可以定义一个 JSON Schema，描述所需 JSON 输出的结构。然后，LLM 将尝试生成符合此模式的 JSON。

    ```json
    {
      "type": "object",
      "properties": {
        "name": {
          "type": "string",
          "description": "The name of the person."
        },
        "age": {
          "type": "integer",
          "description": "The age of the person."
        },
        "city": {
          "type": "string",
          "description": "The city where the person lives."
        }
      },
      "required": ["name", "age"]
    }
    ```

    When you prompt the LLM, you would include this schema in your prompt, instructing it to output a JSON object that matches the schema.

    当您提示 LLM 时，您会在提示中包含此模式，指示它输出一个与该模式匹配的 JSON 对象。

2.  **YAML:** Similar to JSON, you can instruct the LLM to output data in YAML format. YAML is often considered more human-readable than JSON.

    2.  **YAML：** 与 JSON 类似，您可以指示 LLM 以 YAML 格式输出数据。YAML 通常比 JSON 更具人类可读性。

    ```yaml
    name: John Doe
    age: 30
    city: New York
    ```

    You would prompt the LLM with instructions like: "Please provide the information in YAML format."

    您会向 LLM 发出类似以下的提示：“请以 YAML 格式提供信息。”

3.  **Custom Templates:** You can also create custom templates that the LLM should follow. This is useful for more complex or specific output formats.

    3.  **自定义模板：** 您还可以创建 LLM 应遵循的自定义模板。这对于更复杂或特定的输出格式很有用。

    For example, you could ask the LLM to output a list of items, each with a specific set of properties:

    例如，您可以要求 LLM 输出一个项目列表，每个项目都具有一组特定的属性：

    ```
    - item: Apple
      details:
        color: Red
        taste: Sweet
    - item: Banana
      details:
        color: Yellow
        taste: Sweet
    ```

    Your prompt might look like: "Extract the item name and its color and taste. Format the output as a list of items, with each item having a 'details' field containing 'color' and 'taste'."

    您的提示可能看起来像：“提取项目名称及其颜色和口味。将输出格式化为项目列表，每个项目都有一个包含‘颜色’和‘口味’的‘详细信息’字段。”

### Libraries and Tools

### 库和工具

Several libraries and tools can help you implement structured outputs with LLMs:

有几个库和工具可以帮助您使用 LLM 实现结构化输出：

*   **Instructor:** A popular Python library that allows you to easily define Pydantic models and use them to guide LLM output.
*   **LangChain:** Offers various tools and abstractions for working with LLMs, including support for structured output parsing.
*   **LlamaIndex:** Focuses on data integration and querying, and also provides capabilities for structured data handling with LLMs.

*   **Instructor：** 一个流行的 Python 库，允许您轻松定义 Pydantic 模型并使用它们来指导 LLM 输出。
*   **LangChain：** 提供各种用于处理 LLM 的工具和抽象，包括对结构化输出解析的支持。
*   **LlamaIndex：** 专注于数据集成和查询，还提供了使用 LLM 处理结构化数据的能力。

By leveraging these methods and tools, you can significantly improve the reliability and usability of your LLM applications.

通过利用这些方法和工具，您可以显著提高 LLM 应用程序的可靠性和可用性。

````python
from pydantic import BaseModel, Field


class Joke(BaseModel):
    """Joke to tell user."""

    setup: str = Field(description="The setup of the joke")
    punchline: str = Field(description="The punchline to the joke")


structured_llm = llm.with_structured_output(Joke)

structured_llm.invoke("Tell me a joke about cats")
````

## 输入图像

````python
multimodal_llm = ChatSambaNovaCloud(
    model="Llama-3.2-11B-Vision-Instruct",
    max_tokens=1024,
    temperature=0.7,
    top_p=0.01,
)
````

````python
import base64

import httpx

image_url = (
    "https://images.pexels.com/photos/147411/italy-mountains-dawn-daybreak-147411.jpeg"
)
image_data = base64.b64encode(httpx.get(image_url).content).decode("utf-8")

message = HumanMessage(
    content=[
        {"type": "text", "text": "describe the weather in this image in 1 sentence"},
        {
            "type": "image_url",
            "image_url": {"url": f"data:image/jpeg;base64,{image_data}"},
        },
    ],
)
response = multimodal_llm.invoke([message])
print(response.content)
````
