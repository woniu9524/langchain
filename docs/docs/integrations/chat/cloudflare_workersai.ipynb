## 实例化

现在我们可以实例化我们的模型对象并生成聊天补全：

- 使用相关参数更新模型实例化。

````python
from langchain_cloudflare.chat_models import ChatCloudflareWorkersAI

llm = ChatCloudflareWorkersAI(
    model="@cf/meta/llama-3.3-70b-instruct-fp8-fast",
    temperature=0,
    max_tokens=1024,
    # other params...
)
````

## 调用

````python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to French. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = llm.invoke(messages)
ai_msg
````

````python
print(ai_msg.content)
````

## 链式调用

我们可以像这样将模型与提示模板进行[链式调用](/docs/how_to/sequence/)：

````python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate(
    [
        (
            "system",
            "You are a helpful assistant that translates {input_language} to {output_language}.",
        ),
        ("human", "{input}"),
    ]
)

chain = prompt | llm
chain.invoke(
    {
        "input_language": "English",
        "output_language": "German",
        "input": "I love programming.",
    }
)
````

## 结构化输出

This guide explains how to use structured outputs to create more robust and maintainable applications.

本指南将介绍如何使用结构化输出来创建更健壮、更易于维护的应用程序。

Structured outputs allow you to define the exact shape of the data you expect to receive from a model. This makes it easier to parse and validate the model's responses, leading to more reliable applications.

结构化输出允许您定义从模型接收数据的确切形状。这使得解析和验证模型的响应更加容易，从而带来更可靠的应用程序。

### Benefits of Structured Outputs

### 结构化输出的好处

*   **Improved Reliability:** By defining a clear schema, you can ensure that the model's output conforms to your expectations, reducing errors and unexpected behavior.
*   **Easier Data Parsing:** Structured data is much simpler to parse and work with in your application compared to unstructured text.
*   **Enhanced Maintainability:** As your application evolves, a well-defined output structure makes it easier to update and manage your code.
*   **Better Validation:** You can easily validate the model's output against your defined schema, catching potential issues early.

*   **提高可靠性：** 通过定义清晰的模式，您可以确保模型的输出符合您的预期，从而减少错误和意外行为。
*   **更轻松的数据解析：** 与非结构化文本相比，结构化数据在您的应用程序中更容易解析和处理。
*   **增强可维护性：** 随着应用程序的演进，明确定义的输出结构可以更轻松地更新和管理代码。
*   **更好的验证：** 您可以轻松地根据定义的模式验证模型的输出，及早发现潜在问题。

### How to Use Structured Outputs

### 如何使用结构化输出

Most AI platforms provide ways to specify the desired output format. This often involves defining a JSON schema or using specific annotations in your prompts.

大多数人工智能平台都提供了指定所需输出格式的方法。这通常涉及定义一个 JSON 模式或在您的提示中使用特定的注解。

Here's a general approach:

以下是一种通用方法：

1.  **Define Your Schema:** Create a JSON schema that describes the structure and data types of the information you want the model to return.

    1.  **定义您的模式：** 创建一个 JSON 模式，描述您希望模型返回的信息的结构和数据类型。
2.  **Prompt the Model:** Include instructions in your prompt that tell the model to return data in the specified format. You might also provide the schema itself within the prompt.

    2.  **提示模型：** 在您的提示中包含说明，告诉模型以指定的格式返回数据。您也可以在提示中提供模式本身。
3.  **Parse and Validate:** In your application code, parse the model's response according to the defined schema and validate its structure and data types.

    3.  **解析和验证：** 在您的应用程序代码中，根据定义的模式解析模型的响应，并验证其结构和数据类型。

#### Example: Using JSON Schema

#### 示例：使用 JSON 模式

Let's say you want to extract information about a product, including its name, price, and availability.

假设您想提取有关产品的信息，包括其名称、价格和可用性。

**1. Define the JSON Schema:**

**1. 定义 JSON 模式：**

```json
{
  "type": "object",
  "properties": {
    "product_name": {
      "type": "string",
      "description": "The name of the product."
    },
    "price": {
      "type": "number",
      "description": "The price of the product."
    },
    "is_available": {
      "type": "boolean",
      "description": "Whether the product is currently available."
    }
  },
  "required": ["product_name", "price", "is_available"]
}
```

**2. Prompt the Model:**

**2. 提示模型：**

```
Extract the following information about the product mentioned in the text:
- Product Name
- Price
- Availability (true/false)

Return the information in JSON format according to the following schema:
{
  "type": "object",
  "properties": {
    "product_name": {
      "type": "string",
      "description": "The name of the product."
    },
    "price": {
      "type": "number",
      "description": "The price of the product."
    },
    "is_available": {
      "type": "boolean",
      "description": "Whether the product is currently available."
    }
  },
  "required": ["product_name", "price", "is_available"]
}

Text: "The new SuperWidget is now available for $49.99. Get yours today!"
```

**3. Expected Model Output:**

**3. 预期的模型输出：**

```json
{
  "product_name": "SuperWidget",
  "price": 49.99,
  "is_available": true
}
```

**4. Parsing and Validation in Code (Conceptual Example):**

**4. 代码中的解析和验证（概念示例）：**

```javascript
// Assuming 'modelResponse' is the JSON string received from the model
const schema = {
  "type": "object",
  "properties": {
    "product_name": {
      "type": "string",
      "description": "The name of the product."
    },
    "price": {
      "type": "number",
      "description": "The price of the product."
    },
    "is_available": {
      "type": "boolean",
      "description": "Whether the product is currently available."
    }
  },
  "required": ["product_name", "price", "is_available"]
};

try {
  const data = JSON.parse(modelResponse);

  // Basic validation (more robust validation can be done with libraries like Ajv)
  if (
    typeof data.product_name === 'string' &&
    typeof data.price === 'number' &&
    typeof data.is_available === 'boolean' &&
    data.hasOwnProperty('product_name') &&
    data.hasOwnProperty('price') &&
    data.hasOwnProperty('is_available')
  ) {
    console.log("Product Name:", data.product_name);
    console.log("Price:", data.price);
    console.log("Availability:", data.is_available);
  } else {
    console.error("Invalid data structure received.");
  }
} catch (error) {
  console.error("Failed to parse JSON:", error);
}
```

### Using Specific Libraries or Frameworks

### 使用特定库或框架

Many AI development frameworks and libraries offer built-in support for structured outputs, simplifying the process. These tools often handle the prompt engineering and response parsing for you.

许多人工智能开发框架和库都为结构化输出提供了内置支持，从而简化了该过程。这些工具通常会为您处理提示工程和响应解析。

For example, some libraries allow you to define Pydantic models (in Python) or TypeScript interfaces that the AI model should adhere to. The library then automatically generates the necessary prompts and parses the output into your defined data structures.

例如，一些库允许您定义 AI 模型应遵循的 Pydantic 模型（在 Python 中）或 TypeScript 接口。然后，该库会自动生成必要的提示，并将输出解析到您定义的数据结构中。

Consult the documentation of your chosen AI platform or library for specific instructions on how to implement structured outputs.

请参阅您选择的人工智能平台或库的文档，了解有关如何实现结构化输出的具体说明。

````python
json_schema = {
    "title": "joke",
    "description": "Joke to tell user.",
    "type": "object",
    "properties": {
        "setup": {
            "type": "string",
            "description": "The setup of the joke",
        },
        "punchline": {
            "type": "string",
            "description": "The punchline to the joke",
        },
        "rating": {
            "type": "integer",
            "description": "How funny the joke is, from 1 to 10",
            "default": None,
        },
    },
    "required": ["setup", "punchline"],
}
structured_llm = llm.with_structured_output(json_schema)

structured_llm.invoke("Tell me a joke about cats")
````

## 绑定工具

This section describes how to use the `bind` tool to bind a service.
本节介绍如何使用 `bind` 工具来绑定服务。

**Note:** The `bind` tool is deprecated and will be removed in a future release.
**注意：** `bind` 工具已被弃用，将在未来的版本中移除。

You can use the `bind` tool to bind a service to a namespace.
您可以使用 `bind` 工具将服务绑定到命名空间。

```bash
bind <namespace> <service>
```

For example, to bind the `redis` service to the `default` namespace:
例如，将 `redis` 服务绑定到 `default` 命名空间：

```bash
bind default redis
```

This will create a new service binding in the `default` namespace that points to the `redis` service.
这将会在 `default` 命名空间中创建一个新的服务绑定，指向 `redis` 服务。

You can also use the `bind` tool to unbind a service from a namespace.
您也可以使用 `bind` 工具将服务从命名空间中解绑。

```bash
unbind <namespace> <service>
```

For example, to unbind the `redis` service from the `default` namespace:
例如，将 `redis` 服务从 `default` 命名空间解绑：

```bash
unbind default redis
```

This will remove the service binding from the `default` namespace.
这将从 `default` 命名空间中移除服务绑定。
