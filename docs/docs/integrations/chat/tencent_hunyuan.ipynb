## 使用 ChatHunyuan 进行流式输出

This guide shows you how to use ChatHunyuan with streaming to get responses from the model.

本指南将向您展示如何使用 ChatHunyuan 的流式输出功能来获取模型的响应。

### 先决条件

*   **Node.js：** 请确保您的计算机上已安装 Node.js。您可以从 [nodejs.org](https://nodejs.org/) 下载。

首先，安装必要的软件包：

### Basic Streaming Example

### 基本流式输出示例

The following code snippet demonstrates how to use ChatHunyuan with streaming.

以下代码片段演示了如何使用 ChatHunyuan 进行流式输出。

```javascript
import { GoogleGenerativeAI } from "@google/generative-ai";

async function run() {
  // Access your API key as an environment variable (see "Set up your API key" above)
  // For example: const GOOGLE_API_KEY=YOUR_API_KEY
  const MODEL_NAME = "gemini-1.5-flash-latest"; // Or use "gemini-1.5-pro-latest"

  const genAI = new GoogleGenerativeAI(process.env.GOOGLE_API_KEY);
  const model = genAI.getGenerativeModel({ model: MODEL_NAME });

  const chat = model.startChat({
    history: [
      {
        role: "user",
        parts: [{ text: "Hello, I would like to become a professional programmer." }],
      },
      {
        role: "model",
        parts: [{ text: "That's a great ambition! Programming requires dedication and continuous learning. What area of programming interests you the most?" }],
      },
    ],
    generationConfig: {
      maxOutputTokens: 800,
    },
  });

  const result = await chat.sendMessageStream("Tell me more about what you said in your last response.");
  const stream = result.stream;

  for await (const chunk of stream) {
    const chunkText = chunk.text();
    console.log("chunk: ", chunkText);
    // You can process each chunk here as it arrives.
    // For example, you could append it to a UI element.
  }

  console.log("Streaming finished.");
}

run();
```

### Explanation

### 说明

1.  **Import `GoogleGenerativeAI`:** We import the necessary class from the `@google/generative-ai` library.
    1.  **导入 `GoogleGenerativeAI`：** 我们从 `@google/generative-ai` 库导入必要的类。

2.  **Set up API Key and Model:**
    *   Replace `"YOUR_API_KEY"` with your actual API key. It's recommended to store your API key in an environment variable for security.
    *   Specify the `MODEL_NAME`. You can use `"gemini-1.5-flash-latest"` for faster responses or `"gemini-1.5-pro-latest"` for more advanced capabilities.
    1.  **设置 API 密钥和模型：**
        *   将 `"YOUR_API_KEY"` 替换为您的实际 API 密钥。建议将 API 密钥存储在环境变量中以确保安全。
        *   指定 `MODEL_NAME`。您可以使用 `"gemini-1.5-flash-latest"` 来获得更快的响应，或使用 `"gemini-1.5-pro-latest"` 来获得更高级的功能。

3.  **Start a Chat:**
    *   `model.startChat()` initializes a chat session. You can provide `history` to give the model context.
    *   `generationConfig` allows you to set parameters like `maxOutputTokens`.
    1.  **开始聊天：**
        *   `model.startChat()` 初始化一个聊天会话。您可以提供 `history` 来为模型提供上下文。
        *   `generationConfig` 允许您设置 `maxOutputTokens` 等参数。

4.  **Send Message with Streaming:**
    *   `chat.sendMessageStream()` sends a message to the model and returns a stream of response chunks.
    *   The `for await...of` loop iterates over the stream, processing each `chunk` as it arrives.
    *   `chunk.text()` extracts the text content from the chunk.
    1.  **发送带流式输出的消息：**
        *   `chat.sendMessageStream()` 将消息发送给模型，并返回响应块的流。
        *   `for await...of` 循环遍历流，在每个 `chunk` 到达时进行处理。
        *   `chunk.text()` 从块中提取文本内容。

5.  **Process Chunks:** Inside the loop, you can handle each `chunk` as needed. This example simply logs the text to the console. In a real application, you might update a UI element or perform other actions.
    1.  **处理块：** 在循环内部，您可以根据需要处理每个 `chunk`。此示例仅将文本记录到控制台。在实际应用中，您可能会更新 UI 元素或执行其他操作。

### Handling Errors

### 处理错误

It's good practice to wrap your streaming logic in a `try...catch` block to handle potential errors.

将您的流式输出逻辑包装在 `try...catch` 块中以处理潜在错误是一种好习惯。

```javascript
import { GoogleGenerativeAI } from "@google/generative-ai";

  const model = genAI.getGenerativeModel({ model: MODEL_NAME });

  const chat = model.startChat({
    history: [
      {
        role: "user",
        parts: [{ text: "Hello, I would like to become a professional programmer." }],
      },
      {
        role: "model",
        parts: [{ text: "That's a great ambition! Programming requires dedication and continuous learning. What area of programming interests you the most?" }],
      },
    ],
    generationConfig: {
      maxOutputTokens: 800,
    },
  });

  try {
    const result = await chat.sendMessageStream("Tell me more about what you said in your last response.");
    const stream = result.stream;

    for await (const chunk of stream) {
      const chunkText = chunk.text();

    }

    console.error("Error during streaming:", error);
    // Handle the error appropriately, e.g., display an error message to the user.
  }
}

run();
```
