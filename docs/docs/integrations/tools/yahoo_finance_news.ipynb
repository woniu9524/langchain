## 链式示例

````python
import os

os.environ["OPENAI_API_KEY"] = ".."
````

````python
from langchain_community.tools.yahoo_finance_news import YahooFinanceNewsTool
from langgraph.prebuilt import create_react_agent

tools = [YahooFinanceNewsTool()]
agent = create_react_agent("openai:gpt-4.1-mini", tools)
````

````python
input_message = {
    "role": "user",
    "content": "What happened today with Microsoft stocks?",
}

for step in agent.stream(
    {"messages": [input_message]},
    stream_mode="values",
):
    step["messages"][-1].pretty_print()
````

````python
input_message = {
    "role": "user",
    "content": "How does Microsoft feels today comparing with Nvidia?",
}

for step in agent.stream(
    {"messages": [input_message]},
    stream_mode="values",
):
    step["messages"][-1].pretty_print()
````

# YahooFinanceNewsTool 如何工作？

This tool allows you to get the latest news about a specific stock.

该工具允许您获取特定股票的最新新闻。

```python
from langchain_community.tools import YahooFinanceNewsTool

tool = YahooFinanceNewsTool()
```

```python
# Example of using the tool
tool.run("AAPL")
```

This tool takes a stock ticker symbol as input and returns a list of news articles related to that stock. Each article is represented as a dictionary with the following keys:

该工具以股票代码作为输入，并返回与该股票相关的文章列表。每篇文章都表示为一个字典，包含以下键：

*   `title`: The title of the news article.
    *   `title`: 新闻文章的标题。
*   `publisher`: The publisher of the news article.
    *   `publisher`: 新闻文章的发布者。
*   `link`: The URL of the news article.
    *   `link`: 新闻文章的 URL。
*   `published_date`: The date the news article was published.
    *   `published_date`: 新闻文章的发布日期。
