## 通过 JinaAI API 使用 Jina 嵌入模型嵌入文本和查询

JinaAI API 提供了对 Jina 嵌入模型的访问权限，使您能够轻松地将文本和查询转换为向量表示。这些向量可以用于各种下游任务，例如语义搜索、文本分类和聚类。

### 准备工作

在开始之前，请确保您已安装 Jina。

```bash
pip install jina
```

您还需要一个 JinaAI API 密钥。您可以从 [JinaAI 网站](https://jinaai.com/) 获取一个。

### 使用 JinaAI API 嵌入文本

以下是如何使用 JinaAI API 嵌入文本的示例：

```python
from jina.proto import jina_pb2
from jina.clients import Assistant

# 替换为您的 JinaAI API 密钥
api_key = "YOUR_JINAAI_API_KEY"

# 要嵌入的文本
text = "这是一个示例文本。"

# 初始化 Assistant 客户端
assistant = Assistant(api_key=api_key)

# 调用 embed endpoint
response = assistant.embed(text=text)

# 打印嵌入向量
print(response.embeddings)
```

在此示例中，我们首先导入必要的类。然后，我们定义我们的 JinaAI API 密钥和要嵌入的文本。接下来，我们初始化 `Assistant` 客户端，并使用 `embed` 方法调用 API。最后，我们打印返回的嵌入向量。

### 使用 JinaAI API 嵌入查询

您还可以使用 JinaAI API 嵌入查询。查询通常比文本短，并且用于搜索与查询相似的文档。

```python
from jina.proto import jina_pb2
from jina.clients import Assistant

# 替换为您的 JinaAI API 密钥
api_key = "YOUR_JINAAI_API_KEY"

# 要嵌入的查询
query = "示例文本是什么？"

# 初始化 Assistant 客户端
assistant = Assistant(api_key=api_key)

# 调用 embed endpoint
response = assistant.embed(query=query)

# 打印嵌入向量
print(response.embeddings)
```

此示例与嵌入文本的示例非常相似，唯一的区别是我们传递了一个 `query` 参数而不是 `text` 参数。

### 选择嵌入模型

JinaAI API 支持多种嵌入模型。您可以通过 `model` 参数指定要使用的模型。

```python
from jina.proto import jina_pb2
from jina.clients import Assistant

# 替换为您的 JinaAI API 密钥
api_key = "YOUR_JINAAI_API_KEY"

# 要嵌入的文本
text = "这是一个示例文本。"

# 初始化 Assistant 客户端
assistant = Assistant(api_key=api_key)

# 调用 embed endpoint，指定模型
response = assistant.embed(text=text, model="jina-v1")

# 打印嵌入向量
print(response.embeddings)
```

在此示例中，我们将 `model` 参数设置为 `"jina-v1"`，这是 Jina 提供的默认嵌入模型。您可以在 JinaAI 文档中找到可用模型的列表。

### 批量嵌入

对于性能优化，您还可以一次嵌入多个文本或查询。

```python
from jina.proto import jina_pb2
from jina.clients import Assistant

# 替换为您的 JinaAI API 密钥
api_key = "YOUR_JINAAI_API_KEY"

# 要嵌入的文本列表
texts = ["这是第一个示例文本。", "这是第二个示例文本。"]

# 初始化 Assistant 客户端
assistant = Assistant(api_key=api_key)

# 调用 embed endpoint
response = assistant.embed(texts=texts)

# 打印嵌入向量
for embedding in response.embeddings:
    print(embedding)
```

在此示例中，我们将文本列表传递给 `texts` 参数。API 将返回一个嵌入向量列表，每个向量对应输入列表中的一个文本。

### 结论

JinaAI API 提供了一种简单有效的方式来利用 Jina 嵌入模型。通过将文本和查询转换为向量，您可以为各种自然语言处理任务解锁强大的功能。

## 使用 JinaAI API 通过 Jina CLIP 嵌入图像和查询

Jina CLIP 是一个基于 CLIP 的模型，可以让你通过文本描述来搜索图像。JinaAI API 提供了访问 Jina CLIP 的便捷方式。

### 先决条件

- **Python 3.7+**
- **Jina 3.0+**
- **JinaAI 账户和 API 密钥**

### 安装 JinaAI Python SDK

```bash
pip install jinaai
```

### 嵌入图像

要嵌入图像，你需要将图像的 URL 或本地文件路径传递给 `jinaai.clip.embed_image()` 方法。

```python
from jinaai import JinaAI

jinaai = JinaAI(api_key="YOUR_API_KEY")

# 使用图像 URL
image_url = "https://example.com/image.jpg"
embedding = jinaai.clip.embed_image(url=image_url)

# 使用本地图像文件
# embedding = jinaai.clip.embed_image(file_path="/path/to/your/image.jpg")

print(embedding)
```

`embedding` 将是一个包含图像嵌入的列表。

### 嵌入文本查询

要嵌入文本查询，你需要将文本传递给 `jinaai.clip.embed_text()` 方法。

```python
from jinaai import JinaAI

jinaai = JinaAI(api_key="YOUR_API_KEY")

text_query = "A photo of a cat sitting on a mat"
embedding = jinaai.clip.embed_text(text=text_query)

print(embedding)
```

`embedding` 将是一个包含文本查询嵌入的列表。

### 使用嵌入进行搜索

一旦你有了图像和文本的嵌入，你就可以使用它们来搜索相似的图像。你可以通过计算嵌入之间的余弦相似度来做到这一点。

```python
import numpy as np

def cosine_similarity(embedding1, embedding2):
    return np.dot(embedding1, embedding2) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))

# 假设你已经有了 image_embedding 和 text_embedding
similarity_score = cosine_similarity(image_embedding[0], text_embedding[0])

print(f"Similarity score: {similarity_score}")
```

得分越高表示图像和文本查询越相似。

### 示例：搜索相似图像

以下是一个完整的示例，演示了如何嵌入图像和文本，然后搜索最相似的图像：

```python
from jinaai import JinaAI
import numpy as np

jinaai = JinaAI(api_key="YOUR_API_KEY")

# 图像库
image_urls = [
    "https://example.com/cat.jpg",
    "https://example.com/dog.jpg",
    "https://example.com/bird.jpg"
]

# 嵌入图像库中的所有图像
image_embeddings = []
for url in image_urls:
    embedding = jinaai.clip.embed_image(url=url)
    image_embeddings.append(embedding[0])

# 文本查询
text_query = "A photo of a cat"

# 嵌入文本查询
text_embedding = jinaai.clip.embed_text(text=text_query)[0]

# 计算相似度并找到最相似的图像
best_match_index = -1
highest_similarity = -1

for i, img_embedding in enumerate(image_embeddings):
    similarity = cosine_similarity(img_embedding, text_embedding)
    if similarity > highest_similarity:
        highest_similarity = similarity
        best_match_index = i

print(f"The most similar image is: {image_urls[best_match_index]}")

````python
multimodal_embeddings = JinaEmbeddings(jina_api_key="jina_*", model_name="jina-clip-v1")
````

````python
image = "https://avatars.githubusercontent.com/u/126733545?v=4"

description = "Logo of a parrot and a chain on green background"

im = Image.open(requests.get(image, stream=True).raw)
print("Image:")
display(im)
````

````python
image_result = multimodal_embeddings.embed_images([image])
````
