## 回调处理程序

````python
from langchain_community.callbacks import ArthurCallbackHandler
from langchain_core.callbacks import StreamingStdOutCallbackHandler
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
````

创建带有 Arthur 回调处理器的 Langchain LLM

```md
Langchain 是一个用于开发由语言模型驱动的应用程序的框架。它允许您将大型语言模型 (LLM) 与其他计算资源或知识相结合，以实现更强大的应用程序。

Arthur 是一个用于监控和评估 LLM 的平台。它提供了一个回调处理器，可以与 Langchain 集成，以捕获和分析 LLM 的交互。

在本教程中，您将学习如何创建与 Arthur 回调处理器集成的 Langchain LLM。

**先决条件**

* Python 3.7 或更高版本
* Langchain 已安装 (`pip install langchain`)
* Arthur 已安装 (`pip install arthur-client`)

**步骤 1：设置 Arthur**

首先，您需要设置 Arthur。您可以按照 Arthur 的官方文档进行设置。

**步骤 2：创建 Langchain LLM**

接下来，您将创建一个 Langchain LLM。您可以使用任何您喜欢的 LLM，例如 OpenAI、Cohere 或 Hugging Face。

以下是一个使用 OpenAI LLM 的示例：

```python
from langchain.llms import OpenAI
from langchain.callbacks import ArthurCallbackHandler

# 初始化 OpenAI LLM
llm = OpenAI(model_name="text-davinci-003", temperature=0.7)

# 初始化 Arthur 回调处理器
arthur_callback = ArthurCallbackHandler(
    project_id="YOUR_PROJECT_ID",  # 替换为您的 Arthur 项目 ID
    api_key="YOUR_API_KEY"      # 替换为您的 Arthur API 密钥
)

# 将回调处理器添加到 LLM 中
llm.add_callback(arthur_callback)
```

**步骤 3：使用 Langchain LLM**

现在您已经创建了与 Arthur 回调处理器集成的 Langchain LLM，您可以像往常一样使用它。Arthur 将自动捕获和分析 LLM 的交互。

以下是一个使用 LLM 的示例：

```python
response = llm("What is the capital of France?")
print(response)
```

**步骤 4：在 Arthur 中查看结果**

您可以在 Arthur 平台中查看 LLM 的交互。Arthur 将提供有关您的 LLM 的性能、成本和使用的见解。

**结论**

通过将 Langchain LLM 与 Arthur 回调处理器集成，您可以轻松地监控和评估您的 LLM 应用程序。这使您能够识别潜在问题，优化性能，并确保您的应用程序按预期运行。

````python
def make_langchain_chat_llm():
    return ChatOpenAI(
        streaming=True,
        temperature=0.1,
        callbacks=[
            StreamingStdOutCallbackHandler(),
            ArthurCallbackHandler.from_credentials(
                arthur_model_id, arthur_url=arthur_url, arthur_login=arthur_login
            ),
        ],
    )
````

````python
chatgpt = make_langchain_chat_llm()
````

使用此 `run` 函数运行聊天 LLM 会将聊天记录保存在一个持续的列表中，以便对话可以引用先前的消息并将每个响应记录到 Arthur 平台。您可以在[模型仪表板页面](https://app.arthur.ai/)上查看此模型推理的历史记录。

输入 `q` 退出运行循环

````python
def run(llm):
    history = []
    while True:
        user_input = input("\n>>> input >>>\n>>>: ")
        if user_input == "q":
            break
        history.append(HumanMessage(content=user_input))
        history.append(llm(history))
````

````python
run(chatgpt)
````
