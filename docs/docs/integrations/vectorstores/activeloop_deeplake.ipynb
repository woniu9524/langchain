# Activeloop Deep Lake

>[Activeloop Deep Lake](https://docs.deeplake.ai/) 是一个多模态向量存储，用于存储嵌入及其元数据，包括文本、JSON、图像、音频、视频等。它可以将数据保存在本地、您的云端或 Activeloop 存储中。它支持包括嵌入及其属性在内的混合搜索。

本笔记本展示了与 `Activeloop Deep Lake` 相关的基本功能。虽然 `Deep Lake` 可以存储嵌入，但它能够存储任何类型的数据。它是一个无服务器的数据湖，具有版本控制、查询引擎和流式数据加载器，可用于深度学习框架。

更多信息，请参阅 Deep Lake [文档](https://docs.deeplake.ai/)

## 在本地使用 Deep Lake

Deep Lake 允许您在本地计算机上存储和管理数据集。

### 先决条件

* **Python 3.7+**
* **pip**

### 安装

```bash
pip install deeplake
```

### 设置本地存储

Deep Lake 使用一个名为 `~/.deeplake` 的目录来存储所有本地数据集。

您可以通过设置 `DEEPLAKE_HOME` 环境变量来更改此目录：

```bash
export DEEPLAKE_HOME=/path/to/your/deeplake/storage
```

### 创建一个本地数据集

要创建一个本地数据集，请使用 `deeplake.empty()` 函数并指定一个本地路径：

```python
import deeplake

# 创建一个名为 'my_local_dataset' 的本地数据集
ds = deeplake.empty("hub://my_local_dataset")
```

这将创建一个名为 `my_local_dataset` 的新数据集，并将其存储在您的本地 Deep Lake 存储目录中。

### 向本地数据集添加数据

您可以使用 `ds.append()` 方法向数据集添加数据：

```python
# 添加一个文本样本
ds.append({"text": "Hello, Deep Lake!"})

# 添加一个图像样本（假设您有一个名为 'image.png' 的图像文件）
# ds.append({"image": deeplake.Image(path="image.png")})
```

### 访问本地数据集

您可以使用 `deeplake.load()` 函数加载本地数据集：

```python
# 加载名为 'my_local_dataset' 的本地数据集
ds = deeplake.load("hub://my_local_dataset")

# 访问数据集中的样本
print(ds[0])
```

### 本地数据集的优势

* **隐私和安全：** 数据保留在您的本地计算机上，无需上传到云端。
* **离线访问：** 您可以在没有互联网连接的情况下访问和处理数据集。
* **成本效益：** 无需支付云存储费用。
* **低延迟：** 本地访问数据可以实现更快的读写速度。

### 注意事项

* **存储空间：** 确保您的本地计算机有足够的存储空间来容纳数据集。
* **备份：** 建议定期备份您的本地数据集，以防数据丢失。
* **性能：** 对于非常大的数据集，本地存储的性能可能会受到您的计算机硬件的限制。

````python
from langchain_deeplake.vectorstores import DeeplakeVectorStore
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
````

````python
import getpass
import os

if "OPENAI_API_KEY" not in os.environ:
    os.environ["OPENAI_API_KEY"] = getpass.getpass("OpenAI API Key:")

if "ACTIVELOOP_TOKEN" not in os.environ:
    os.environ["ACTIVELOOP_TOKEN"] = getpass.getpass("activeloop token:")
````

````python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../../how_to/state_of_the_union.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)

embeddings = OpenAIEmbeddings()
````

### 创建本地数据集

在 `./my_deeplake/` 下创建数据集，然后运行相似性搜索。Deeplake+LangChain 集成在底层使用了 Deep Lake 数据集，因此 `dataset` 和 `vector store` 可以互换使用。要在您自己的云或 Deep Lake 存储中创建数据集，请[相应地调整路径](https://docs.deeplake.ai/latest/getting-started/storage-and-creds/storage-options/)。

````python
db = DeeplakeVectorStore(
    dataset_path="./my_deeplake/", embedding_function=embeddings, overwrite=True
)
db.add_documents(docs)
# or shorter
# db = DeepLake.from_documents(docs, dataset_path="./my_deeplake/", embedding_function=embeddings, overwrite=True)
````

### 查询数据集

````python
query = "What did the president say about Ketanji Brown Jackson"
docs = db.similarity_search(query)
````

````python
print(docs[0].page_content)
````

稍后，您可以在不重新计算嵌入的情况下重新加载数据集

````python
db = DeeplakeVectorStore(
    dataset_path="./my_deeplake/", embedding_function=embeddings, read_only=True
)
docs = db.similarity_search(query)
````

将 `read_only=True` 设置为 `True` 可防止在不需要更新时意外修改向量存储。这可确保数据保持不变，除非明确意图。通常，指定此参数以避免意外更新是一种好习惯。

### 基于属性的元数据过滤

让我们创建另一个包含元数据的向量存储，其中包含文档创建的年份。

````python
import random

for d in docs:
    d.metadata["year"] = random.randint(2012, 2014)

db = DeeplakeVectorStore.from_documents(
    docs, embeddings, dataset_path="./my_deeplake/", overwrite=True
)
````

````python
db.similarity_search(
    "What did the president say about Ketanji Brown Jackson",
    filter={"metadata": {"year": 2013}},
)
````

### TQL 搜索

TQL (Temporal Query Language) is a powerful tool for querying data stored in a time-series database. It allows you to specify conditions and retrieve data points that match those conditions.

TQL（时间序列查询语言）是一个强大的工具，用于查询存储在时间序列数据库中的数据。它允许您指定条件并检索匹配这些条件的数​​据点。

Here are some common TQL search patterns:

以下是一些常见的 TQL 搜索模式：

#### 1. Basic Metric Search

This is the most fundamental search. You specify a metric name and optionally a time range.

#### 1. 基本指标搜索

这是最基础的搜索。您指定一个指标名称，并可选择指定一个时间范围。

```tql
cpu.usage
```

This query retrieves all data points for the `cpu.usage` metric.

此查询检索 `cpu.usage` 指标的所有数据点。

To specify a time range, you can use the `from` and `to` keywords:

要指定时间范围，您可以使用 `from` 和 `to` 关键字：

```tql
cpu.usage from 1h ago to now
```

This query retrieves `cpu.usage` data points from the last hour.

此查询检索过去一小时的 `cpu.usage` 数据点。

#### 2. Filtering by Tags

Time-series data is often associated with tags, which are key-value pairs that provide additional context. You can filter your search based on these tags.

#### 2. 按标签过滤

时间序列数据通常与标签相关联，标签是提供额外上下文的键值对。您可以根据这些标签过滤搜索。

```tql
cpu.usage{host=server1, region=us-east-1}
```

This query retrieves `cpu.usage` data points specifically from `server1` in the `us-east-1` region.

此查询专门检索来自 `us-east-1` 区域的 `server1` 的 `cpu.usage` 数据点。

You can use various operators for tag matching:

您可以使用各种运算符进行标签匹配：

*   `=` (equals)
*   `!=` (not equals)
*   `=~` (regex match)
*   `!~` (regex not match)

*   `=` (等于)
*   `!=` (不等于)
*   `=~` (正则表达式匹配)
*   `!~` (正则表达式不匹配)

```tql
cpu.usage{host!~server[0-9]+}
```

This query retrieves `cpu.usage` data points from hosts that do not match the pattern `server[0-9]+`.

此查询检索不匹配模式 `server[0-9]+` 的主机的 `cpu.usage` 数据点。

#### 3. Aggregations

Often, you don't want raw data points but rather an aggregated view over a time range. TQL supports various aggregation functions.

#### 3. 聚合

通常，您不想要原始数据点，而是想要一个时间范围内的聚合视图。TQL 支持各种聚合函数。

Common aggregation functions include:

常见的聚合函数包括：

*   `sum()`
*   `avg()`
*   `min()`
*   `max()`
*   `count()`
*   `percentile(0.95)`

*   `sum()`
*   `avg()`
*   `min()`
*   `max()`
*   `count()`
*   `percentile(0.95)`

```tql
avg(cpu.usage) from 1h ago to now
```

This query calculates the average `cpu.usage` over the last hour.

此查询计算过去一小时的平均 `cpu.usage`。

You can also group by tags to get aggregated values for different tag combinations:

您还可以按标签分组以获取不同标签组合的聚合值：

```tql
avg(cpu.usage) by host from 1h ago to now
```

This query calculates the average `cpu.usage` for each host over the last hour.

此查询计算过去一小时内每个主机的平均 `cpu.usage`。

#### 4. Combining Metrics (Functions)

TQL allows you to perform operations on multiple metrics or apply functions to a single metric.

#### 4. 组合指标（函数）

TQL 允许您对多个指标执行操作或将函数应用于单个指标。

```tql
rate(http.requests.total)
```

This query calculates the per-second rate of change for the `http.requests.total` counter.

此查询计算 `http.requests.total` 计数器的每秒变化率。

```tql
sum(cpu.usage) + sum(mem.usage)
```

This query calculates the sum of `cpu.usage` and `mem.usage`.

此查询计算 `cpu.usage` 和 `mem.usage` 的总和。

#### 5. Using `fill` for Missing Data

When aggregating or performing calculations, you might encounter gaps in your data. The `fill` clause allows you to specify how to handle these gaps.

#### 5. 使用 `fill` 处理缺失数据

在聚合或执行计算时，您可能会遇到数据中的间隙。`fill` 子句允许您指定如何处理这些间隙。

```tql
avg(cpu.usage) from 1h ago to now fill with 0
```

This query calculates the average `cpu.usage` over the last hour, filling any missing data points with `0`.

此查询计算过去一小时的平均 `cpu.usage`，并用 `0` 填充任何缺失的数据点。

Common `fill` options include:

常见的 `fill` 选项包括：

*   `0`
*   `null`
*   `previous`
*   `linear`

*   `0`
*   `null`
*   `previous`
*   `linear`

#### 6. Complex Queries

You can combine all the above concepts to build complex queries.

#### 6. 复杂查询

您可以组合以上所有概念来构建复杂的查询。

```tql
avg(rate(http.requests.total) by host) from 1d ago to now where region = "us-west-2" and status_code = 500
```

This query calculates the average per-second rate of HTTP requests for each host in the `us-west-2` region that returned a `500` status code, over the last day.

此查询计算过去一天内，在 `us-west-2` 区域返回了 `500` 状态码的每个主机的 HTTP 请求的平均每秒速率。

TQL is a versatile language, and exploring its capabilities will allow you to gain deeper insights from your time-series data.

TQL 是一种多功能语言，探索其功能将使您能够从时间序列数据中获得更深入的见解。

此外，`similarity_search` 方法还支持查询的执行，您可以使用 Deep Lake 的张量查询语言 (TQL) 来指定查询。
