# VSDX

> Visio 文件（扩展名为 .vsdx）与用于创建图表的软件 Microsoft Visio 相关联。它存储了图表的结构、布局和图形元素的信息。此格式有助于在业务、工程和计算机科学等领域创建和共享可视化内容。

Visio 文件可以包含多个页面。其中一些页面可能作为其他页面的背景，这甚至可能跨越多个图层。此 **loader** 会提取每个页面及其关联页面的文本内容，从而能够提取每个页面的所有可见文本，类似于 OCR 算法所做的操作。

**警告**：此加载程序仅兼容扩展名为 **.vsdx** 的 Visio 文件。诸如 .vsd 等扩展名的文件不兼容，因为它们无法转换为压缩的 XML。

```python
from langchain_community.document_loaders import VsdxLoader
```

```python
loader = VsdxLoader(file_path="./example_data/fake.vsdx")
documents = loader.load()
```

**显示已加载的文档**

```python
for i, doc in enumerate(documents):
    print(f"\n------ Page {doc.metadata['page']} ------")
    print(f"Title page : {doc.metadata['page_name']}")
    print(f"Source : {doc.metadata['source']}")
    print("\n==> CONTENT <== ")
    print(doc.page_content)
```
