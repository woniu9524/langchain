# AstraDB

AstraDB 是一个无服务器的、基于 Apache Cassandra 的数据库即服务 (DBaaS)。

AstraDB 旨在为开发人员提供一个易于使用的平台，用于构建和扩展由 Cassandra 提供支持的云原生应用程序。它提供了一个完全托管的数据库，无需管理底层基础设施，并提供了一个简单的 API 来访问和操作数据。

AstraDB 的主要特性包括：

*   **无服务器架构**：AstraDB 的无服务器架构意味着您无需担心数据库的配置、扩展或维护。AstraDB 会自动处理这些任务，让您可以专注于构建应用程序。
*   **基于 Cassandra**：AstraDB 基于 Apache Cassandra，一个高度可扩展的、分布式 NoSQL 数据库。这意味着您可以利用 Cassandra 的强大功能，如高可用性、线性可伸缩性和多区域复制。
*   **云原生**：AstraDB 是为云环境设计的，可以轻松地与各种云服务和工具集成。
*   **简单 API**：AstraDB 提供了一个简单的 RESTful API，用于访问和操作数据。这使得开发人员可以轻松地将 AstraDB 集成到他们的应用程序中，而无需学习复杂的数据库查询语言。
*   **按需付费**：AstraDB 采用按需付费模式，您只需为您使用的资源付费。这使得它成为各种规模的应用程序的经济高效的解决方案。

AstraDB 可用于各种用例，包括：

*   **实时数据应用程序**：AstraDB 的低延迟和高吞吐量使其成为构建实时数据应用程序的理想选择，例如物联网 (IoT) 数据摄取、游戏和金融交易。
*   **个性化和推荐引擎**：AstraDB 的可伸缩性和灵活性使其非常适合存储和检索用户数据，以提供个性化的体验和推荐。
*   **内容管理系统**：AstraDB 可用于存储和管理大量内容，例如文章、视频和图像，并提供快速访问。
*   **微服务**：AstraDB 的无服务器特性和简单的 API 使其成为构建和扩展微服务的理想选择。

**开始使用 AstraDB**

要开始使用 AstraDB，您需要创建一个 AstraDB 账户。您可以访问 [https://astra.datastax.com/](https://astra.datastax.com/) 进行注册。

创建账户后，您可以创建一个新的 AstraDB 数据库。您需要选择数据库的名称、区域和云提供商。

创建数据库后，您将获得一个数据库凭据，您可以使用该凭据通过 API 访问您的数据库。

**AstraDB 的优势**

*   **简化数据库管理**：AstraDB 的无服务器架构消除了数据库管理的复杂性，让开发人员可以专注于应用程序开发。
*   **可伸缩性和性能**：AstraDB 基于 Cassandra，提供卓越的可伸缩性和性能，以满足不断增长的数据需求。
*   **成本效益**：按需付费模式和无服务器架构使 AstraDB 成为一种经济高效的解决方案。
*   **易于集成**：简单的 API 和云原生设计使 AstraDB 易于与现有应用程序和工作流集成。

**AstraDB 的局限性**

*   **学习曲线**：虽然 AstraDB 旨在简化数据库管理，但熟悉 Cassandra 的概念和最佳实践仍然有益。
*   **供应商锁定**：作为一项托管服务，AstraDB 可能会导致供应商锁定，尽管其基于开源 Cassandra 这一事实减轻了这一点。
*   **有限的自定义**：与自托管 Cassandra 相比，AstraDB 的自定义选项可能有限，因为底层基础设施由 DataStax 管理。

总而言之，AstraDB 是一个强大的无服务器数据库即服务，它利用 Apache Cassandra 的能力来提供一个可伸缩、高性能且易于使用的平台，用于构建云原生应用程序。它通过简化数据库管理、提供卓越的性能和成本效益，使开发人员能够专注于构建创新的解决方案。

## 概述

Astra DB 文档加载器会返回从 Astra DB 集合中读取的 Langchain `Document` 对象列表。

加载器接受以下参数：

* `api_endpoint`: Astra DB API 端点。格式类似 `https://01234567-89ab-cdef-0123-456789abcdef-us-east1.apps.astra.datastax.com`
* `token`: Astra DB 令牌。格式类似 `AstraCS:aBcD0123...`
* `collection_name`: AstraDB 集合名称
* `namespace`: (可选) AstraDB 命名空间 (在 Astra DB 中称为 _keyspace_)
* `filter_criteria`: (可选) 在 find 查询中使用的过滤器
* `projection`: (可选) 在 find 查询中使用的投影
* `limit`: (可选) 要检索的文档的最大数量
* `extraction_function`: (可选) 将 AstraDB 文档转换为 LangChain `page_content` 字符串的函数。默认为 `json.dumps`

加载器为读取的文档设置以下元数据：

```python
metadata={
    "namespace": "...",
    "api_endpoint": "...",
    "collection": "..."
}

## 使用文档加载器加载文档

````python
from langchain_astradb import AstraDBLoader
````

[**API 参考:** `AstraDBLoader`](https://python.langchain.com/api_reference/astradb/document_loaders/langchain_astradb.document_loaders.AstraDBLoader.html#langchain_astradb.document_loaders.AstraDBLoader)

````python
from getpass import getpass

ASTRA_DB_API_ENDPOINT = input("ASTRA_DB_API_ENDPOINT = ")
ASTRA_DB_APPLICATION_TOKEN = getpass("ASTRA_DB_APPLICATION_TOKEN = ")
````

````python
loader = AstraDBLoader(
    api_endpoint=ASTRA_DB_API_ENDPOINT,
    token=ASTRA_DB_APPLICATION_TOKEN,
    collection_name="movie_reviews",
    projection={"title": 1, "reviewtext": 1},
    limit=10,
)
````

````python
docs = loader.load()
````
