## 分割模式与自定义页面分隔符

This guide explains how to use the splitting mode and custom page delimiters.
本指南将介绍如何使用分割模式和自定义页面分隔符。

### Splitting mode
### 分割模式

Splitting mode allows you to split your document into multiple pages.
分割模式允许您将文档分割成多个页面。

#### How it works
#### 工作原理

The splitting mode works by looking for a specific delimiter in your document. When it finds the delimiter, it starts a new page.
分割模式通过在文档中查找特定的分隔符来工作。当找到分隔符时，它会开始新的一页。

#### Default delimiter
#### 默认分隔符

The default delimiter is `---`.
默认分隔符是 `---`。

#### Custom delimiter
#### 自定义分隔符

You can also use a custom delimiter. To do this, you need to set the `delimiter` option in the `next.config.js` file.
您也可以使用自定义分隔符。为此，您需要在 `next.config.js` 文件中设置 `delimiter` 选项。

```javascript
// next.config.js
module.exports = {
  // ... other configurations
  experimental: {
    mdxRs: true,
    mdx: {
      // ... other mdx options
      delimiter: '===',
    },
  },
}
```

In this example, the delimiter is set to `===`.
在此示例中，分隔符设置为 `===`。

### Custom pages delimiter
### 自定义页面分隔符

You can also use custom page delimiters to split your document into multiple pages.
您也可以使用自定义页面分隔符将文档分割成多个页面。

#### How it works
#### 工作原理

The custom pages delimiter works by looking for a specific delimiter in your document. When it finds the delimiter, it starts a new page.
自定义页面分隔符通过在文档中查找特定的分隔符来工作。当找到分隔符时，它会开始新的一页。

#### Default delimiter
#### 默认分隔符

The default delimiter is `---`.
默认分隔符是 `---`。

#### Custom delimiter
#### 自定义分隔符

You can also use a custom delimiter. To do this, you need to set the `pageDelimiter` option in the `next.config.js` file.
您也可以使用自定义分隔符。为此，您需要在 `next.config.js` 文件中设置 `pageDelimiter` 选项。

```javascript
// next.config.js
module.exports = {
  // ... other configurations
  experimental: {
    mdxRs: true,
    mdx: {
      // ... other mdx options
      pageDelimiter: '***',
    },
  },
}
```

In this example, the delimiter is set to `***`.
在此示例中，分隔符设置为 `***`。

### Example
### 示例

Here is an example of a document that uses the splitting mode and custom pages delimiter:
以下是一个使用分割模式和自定义页面分隔符的文档示例：

```md
# Page 1
This is the first page.
---
# Page 2
This is the second page.
***
# Page 3
This is the third page.
```

In this example, the document is split into three pages. The first page is separated by `---`, and the second page is separated by `***`.
在此示例中，文档被分割成三个页面。第一个页面由 `---` 分隔，第二个页面由 `***` 分隔。

加载 PDF 文件时，您可以通过两种不同的方式进行拆分：
- 按页拆分
- 作为单一文本流

默认情况下，PyPDFLoader 会将 PDF 作为单一文本流进行拆分。

### 按页提取 PDF。每页提取为一个 langchain Document 对象：

````python
loader = PyPDFLoader(
    "./example_data/layout-parser-paper.pdf",
    mode="page",
)
docs = loader.load()
print(len(docs))
pprint.pp(docs[0].metadata)
````

在此模式下，PDF 将按页拆分，生成的 Documents 元数据包含页码。但在某些情况下，我们可能希望将 PDF 作为单个文本流进行处理（这样就不会将某些段落截断）。在这种情况下，您可以使用 *single* 模式：

### 将整个 PDF 提取为单个 langchain Document 对象：

````python
loader = PyPDFLoader(
    "./example_data/layout-parser-paper.pdf",
    mode="single",
)
docs = loader.load()
print(len(docs))
pprint.pp(docs[0].metadata)
````

在逻辑上，在此模式下，“page_number”元数据会消失。以下是清晰识别文本流中页面结束位置的方法：

### 为 *single* 模式添加自定义的 *pages_delimiter* 以识别页面结束位置：

````python
loader = PyPDFLoader(
    "./example_data/layout-parser-paper.pdf",
    mode="single",
    pages_delimiter="\n-------THIS IS A CUSTOM END OF PAGE-------\n",
)
docs = loader.load()
print(docs[0].page_content[:5780])
````

这可以简单地是 `\n`，或者 `\f` 来明确指示分页符，或者 `<!-- PAGE BREAK -->` 以无视觉效果地无缝注入 Markdown 查看器。

# 从 PDF 中提取图像

您可以通过以下三种解决方案从 PDF 中提取图像：
- rapidOCR（轻量级光学字符识别工具）
- Tesseract（高精度 OCR 工具）
- 多模态语言模型

您可以调整这些函数，选择提取图像的输出格式为 *html*、*markdown* 或 *text*。

结果将插入在页面倒数第二段和倒数第一段文本之间。

### 使用 rapidOCR 从 PDF 中提取图像：

````python
%pip install -qU rapidocr-onnxruntime
````

````python
from langchain_community.document_loaders.parsers import RapidOCRBlobParser

loader = PyPDFLoader(
    "./example_data/layout-parser-paper.pdf",
    mode="page",
    images_inner_format="markdown-img",
    images_parser=RapidOCRBlobParser(),
)
docs = loader.load()

print(docs[5].page_content)
````

请注意，RapidOCR 的设计是为了处理中文和英文，而非其他语言。

### 使用 Tesseract 从 PDF 中提取图像：

````python
%pip install -qU pytesseract
````

````python
from langchain_community.document_loaders.parsers import TesseractBlobParser

loader = PyPDFLoader(
    "./example_data/layout-parser-paper.pdf",
    mode="page",
    images_inner_format="html-img",
    images_parser=TesseractBlobParser(),
)
docs = loader.load()
print(docs[5].page_content)
````

## 处理文件

许多文档加载器都涉及解析文件。这类加载器之间的区别通常源于解析文件的方式，而不是加载文件的方式。例如，您可以使用 `open` 来读取 PDF 或 markdown 文件的二进制内容，但需要不同的解析逻辑才能将该二进制数据转换为文本。

因此，将解析逻辑与加载逻辑解耦会很有帮助，这样可以更轻松地重用给定的解析器，而无需考虑数据是如何加载的。
您可以使用此策略来分析不同的文件，并使用相同的解析参数。

````python
from langchain_community.document_loaders import FileSystemBlobLoader
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.parsers import PyPDFParser

loader = GenericLoader(
    blob_loader=FileSystemBlobLoader(
        path="./example_data/",
        glob="*.pdf",
    ),
    blob_parser=PyPDFParser(),
)
docs = loader.load()
print(docs[0].page_content)
pprint.pp(docs[0].metadata)
````

可以处理来自云存储的文件。

````python
from langchain_community.document_loaders import CloudBlobLoader
from langchain_community.document_loaders.generic import GenericLoader

loader = GenericLoader(
    blob_loader=CloudBlobLoader(
        url="s3://mybucket",  # Supports s3://, az://, gs://, file:// schemes.
        glob="*.pdf",
    ),
    blob_parser=PyPDFParser(),
)
docs = loader.load()
print(docs[0].page_content)
pprint.pp(docs[0].metadata)
````
