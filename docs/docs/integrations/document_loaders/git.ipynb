## 从 URL 克隆仓库

You can clone a repository from a URL using the `git clone` command.

您可以使用 `git clone` 命令从 URL 克隆仓库。

```bash
git clone https://github.com/git/git
```

This command downloads the entire repository history and checks out the latest version of the code.

此命令会下载整个仓库历史并检出代码的最新版本。

**Note:** Replace `https://github.com/git/git` with the URL of the repository you want to clone.

**注意：** 将 `https://github.com/git/git` 替换为您要克隆的仓库的 URL。

You can also clone a specific branch of a repository by using the `-b` flag:

您也可以使用 `-b` 标志克隆仓库的特定分支：

```bash
git clone -b develop https://github.com/git/git
```

This command clones the `develop` branch of the repository.

此命令会克隆仓库的 `develop` 分支。

## 过滤要加载的文件

This guide explains how to filter files to load.

本指南将介绍如何过滤要加载的文件。

You can filter files to load by specifying a `filter` option in the `load` function.

您可以通过在 `load` 函数中指定 `filter` 选项来过滤要加载的文件。

```js
import { load } from 'node:fs';

const files = await load('./data', {
  filter: (filename) => filename.endsWith('.txt')
});

console.log(files);
```

This example loads all files in the `./data` directory that end with `.txt`.

此示例将加载 `./data` 目录中所有以 `.txt` 结尾的文件。

The `filter` option takes a function that accepts a `filename` as an argument and returns `true` if the file should be loaded, and `false` otherwise.

`filter` 选项接受一个函数，该函数以 `filename` 作为参数，如果文件应被加载则返回 `true`，否则返回 `false`。

You can also use the `filter` option to exclude files.

您也可以使用 `filter` 选项来排除文件。

```js
import { load } from 'node:fs';

const files = await load('./data', {
  filter: (filename) => !filename.startsWith('.')
});

console.log(files);
```

This example loads all files in the `./data` directory that do not start with a dot (`.`).

此示例将加载 `./data` 目录中所有不以点号（`.`）开头的文件。

````python
from langchain_community.document_loaders import GitLoader

# e.g. loading only python files
loader = GitLoader(
    repo_path="./example_data/test_repo1/",
    file_filter=lambda file_path: file_path.endswith(".py"),
)
````
