# 响应元数据

许多模型提供商会在其聊天生成[响应](/docs/concepts/messages/#aimessage)中包含一些元数据。此元数据可以通过 `AIMessage.response_metadata: Dict` 属性访问。根据模型提供商和模型配置的不同，其中可能包含[令牌计数](/docs/how_to/chat_token_usage_tracking)、[对数概率](/docs/how_to/logprobs)等信息。

以下是不同提供商的响应元数据示例：

## OpenAI

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")
msg = llm.invoke("What's the oldest known example of cuneiform")
msg.response_metadata
```

## Anthropic

```python
from langchain_anthropic import ChatAnthropic

llm = ChatAnthropic(model="claude-3-5-sonnet-latest")
msg = llm.invoke("What's the oldest known example of cuneiform")
msg.response_metadata
```

*   **模型评估和解释：**
    *   **Vertex AI Model Evaluation：** 评估模型的性能。
    *   **Vertex AI Explainable AI：** 提供模型预测的解释，帮助理解模型行为。
*   **模型部署和提供：**
    *   **Vertex AI Endpoints：** 将训练好的模型部署为可扩展的在线预测服务。
    *   **Vertex AI Batch Prediction：** 对大量数据进行批量预测。
*   **模型监控：**
    *   **Vertex AI Model Monitoring：** 监控已部署模型的性能和数据漂移，并在出现问题时发出警报。
*   **MLOps 工具：**
    *   **Vertex AI Pipelines：** 使用 Kubeflow Pipelines 构建和自动化机器学习工作流。
    *   **Vertex AI Experiments：** 跟踪和比较模型训练实验。
    *   **Vertex AI Model Registry：** 注册、版本化和管理您的模型。

**优势：**

*   **统一平台：** 将机器学习工作流中的所有工具和流程整合到一个平台中。
*   **可扩展性：** 利用 Google Cloud 的基础设施进行大规模数据处理和模型训练。

*   **端到端 MLOps：** 支持整个机器学习生命周期的自动化和管理。

**适用场景：**

Vertex AI 适用于各种机器学习应用，包括：
