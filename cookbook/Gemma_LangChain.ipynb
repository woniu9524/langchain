## 在本地从 HuggingFace 运行 Gemma

You can run Gemma locally using the Hugging Face `transformers` library.

您可以使用 Hugging Face 的 `transformers` 库在本地运行 Gemma。

First, you need to install the necessary libraries:

首先，您需要安装必要的库：

```bash
pip install transformers accelerate bitsandbytes
```

Next, you'll need to accept the Gemma terms of use on the Hugging Face website. You can find the Gemma models here:

接下来，您需要在 Hugging Face 网站上接受 Gemma 的使用条款。您可以在此处找到 Gemma 模型：

*   [gemma-2b](https://huggingface.co/google/gemma-2b)
*   [gemma-7b](https://huggingface.co/google/gemma-7b)

Once you have accepted the terms, you can log in to Hugging Face from your terminal:

接受条款后，您可以从终端登录 Hugging Face：

```bash
huggingface-cli login
```

This will prompt you to enter your Hugging Face authentication token.

这将提示您输入 Hugging Face 身份验证令牌。

Now, you can load and run the model using the following Python code:

现在，您可以使用以下 Python 代码加载和运行模型：

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Specify the model name
model_name = "google/gemma-2b" # or "google/gemma-7b"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    # Uncomment the following line if you have bitsandbytes installed and want to use 4-bit quantization
    # load_in_4bit=True,
)

# Define the prompt
prompt = "What is the capital of France?"

# Tokenize the prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt").to(model.device)

# Generate text
# You can adjust generation parameters like max_new_tokens, temperature, etc.
outputs = model.generate(input_ids, max_new_tokens=50, do_sample=True, temperature=0.7, top_p=0.9)

# Decode and print the output
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```

**Explanation:**

**说明：**

*   `pip install transformers accelerate bitsandbytes`: Installs the necessary libraries. `transformers` is the core library for working with Hugging Face models. `accelerate` helps with distributed training and inference. `bitsandbytes` is used for quantization (reducing model size and memory usage).
*   `huggingface-cli login`: Logs you into your Hugging Face account, allowing you to download gated models like Gemma.
*   `AutoTokenizer.from_pretrained(model_name)`: Loads the tokenizer associated with the specified Gemma model. The tokenizer converts text into numerical tokens that the model can understand.
*   `AutoModelForCausalLM.from_pretrained(...)`: Loads the Gemma model itself.
    *   `torch_dtype=torch.bfloat16`: Specifies the data type for the model weights. `bfloat16` is a half-precision floating-point format that can significantly speed up inference and reduce memory usage compared to `float32`, while maintaining good accuracy.
    *   `device_map="auto"`: Automatically distributes the model across available devices (like GPUs). If you have multiple GPUs, it will use them. If you only have a CPU, it will use that.
    *   `load_in_4bit=True` (commented out): If you uncomment this and have `bitsandbytes` installed, the model will be loaded in 4-bit precision, further reducing memory requirements. This is useful for running larger models on less powerful hardware.
*   `tokenizer.encode(prompt, return_tensors="pt").to(model.device)`: Converts your text prompt into a tensor of input IDs, ensuring it's on the same device as the model.
*   `model.generate(...)`: Generates text based on the input prompt.
    *   `max_new_tokens=50`: Limits the generated text to a maximum of 50 new tokens.
    *   `do_sample=True`: Enables sampling, meaning the model chooses the next token probabilistically rather than always picking the most likely one. This leads to more varied and creative output.
    *   `temperature=0.7`: Controls the randomness of the sampling. Lower values make the output more focused and deterministic, while higher values increase randomness.
    *   `top_p=0.9`: Nucleus sampling parameter. The model considers only the most probable tokens whose cumulative probability exceeds `top_p`.
*   `tokenizer.decode(outputs[0], skip_special_tokens=True)`: Converts the generated token IDs back into human-readable text, skipping any special tokens used by the model.

**Running on GPU:**

**在 GPU 上运行：**

To ensure the model runs on your GPU, make sure you have the correct CUDA drivers and PyTorch version installed. The `device_map="auto"` argument should automatically detect and utilize your GPU if available.

为确保模型在您的 GPU 上运行，请确保您已安装正确的 CUDA 驱动程序和 PyTorch 版本。`device_map="auto"` 参数应在可用时自动检测并利用您的 GPU。

If you encounter memory issues, consider using 4-bit quantization by uncommenting the `load_in_4bit=True` line in the `AutoModelForCausalLM.from_pretrained` call.

如果您遇到内存问题，请考虑通过取消注释 `AutoModelForCausalLM.from_pretrained` 调用中的 `load_in_4bit=True` 行来使用 4 位量化。

````python
llm = GemmaLocalHF(model_name="google/gemma-2b", hf_access_token=hf_access_token)
````

````python
output = llm.invoke("What is the meaning of life?", max_tokens=50)
print(output)
````

与上面相同，在本地使用 Gemma 作为多轮聊天模型。您可能需要重新启动笔记本并清理 GPU 内存以避免 OOM 错误：

````python
llm = GemmaChatLocalHF(model_name=model_name, hf_access_token=hf_access_token)
````

````python
from langchain_core.messages import HumanMessage

message1 = HumanMessage(content="Hi! Who are you?")
answer1 = llm.invoke([message1], max_tokens=60)
print(answer1)
````

````python
message2 = HumanMessage(content="What can you help me with?")
answer2 = llm.invoke([message1, answer1, message2], max_tokens=140)

print(answer2)
````

以及后处理也是如此：

````python
answer1 = llm.invoke([message1], max_tokens=60, parse_response=True)
print(answer1)

answer2 = llm.invoke([message1, answer1, message2], max_tokens=120, parse_response=True)
print(answer2)
````
