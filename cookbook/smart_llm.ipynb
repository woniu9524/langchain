# 如何使用 SmartLLMChain

SmartLLMChain 是一种自我批评链的形式，如果您有特别复杂的问题需要回答，它可以帮助您。它不会只进行一次 LLM 传递，而是执行以下 3 个步骤：
1. 构思：将用户提示通过 LLM 传递 n 次，以获得 n 个输出提案（称为“想法”），其中 n 是您可以设置的参数
2. 批评：LLM 批评所有想法，找出可能的缺陷并选择最佳想法
3. 解决：LLM 尝试改进最佳想法（如批评步骤中所选），并输出它。然后这就是最终输出。

SmartLLMChains 基于 https://youtu.be/wVzuvf9D9BU 中提出的 SmartGPT 工作流。

请注意，SmartLLMChains
- 使用更多的 LLM 传递（即 n+2 次，而不是仅 1 次）
- 仅在底层 LLM 具备反思能力时才有效，而小型模型通常不具备
- 仅适用于返回恰好 1 个输出的底层模型，而不是多个输出

本笔记本演示了如何使用 SmartLLMChain。

##### 所有步骤使用相同的 LLM

```python
import os

os.environ["OPENAI_API_KEY"] = "..."
```

```python
from langchain.prompts import PromptTemplate
from langchain_experimental.smart_llm import SmartLLMChain
from langchain_openai import ChatOpenAI
```

作为示例问题，我们将使用“我有一个 12 升的壶和一个 6 升的壶。我想量出 6 升。我该怎么做？”。这是原始 SmartGPT 视频中的一个示例（https://youtu.be/wVzuvf9D9BU?t=384）。虽然这看起来是一个非常简单的问题，但大型语言模型在处理涉及数字和物理推理的这类问题时会遇到困难。

正如我们将看到的，所有 3 个初步想法都是完全错误的——即使我们使用的是 GPT4！只有在使用自我反思时，我们才能得到正确的答案。

```python
hard_question = "I have a 12 liter jug and a 6 liter jug. I want to measure 6 liters. How do I do it?"
```

所以，我们首先创建一个 LLM 和提示模板

```python
prompt = PromptTemplate.from_template(hard_question)
llm = ChatOpenAI(temperature=0, model_name="gpt-4")
```

现在我们可以创建一个 SmartLLMChain

```python
chain = SmartLLMChain(llm=llm, prompt=prompt, n_ideas=3, verbose=True)
```

现在我们可以使用 SmartLLM 作为我们 LLM 的即插即用替代品。例如：

```python
chain.invoke({})
```

##### 不同的 LLM 用于不同的步骤

您也可以通过传递 `ideation_llm`、`critique_llm` 和 `resolve_llm` 来为不同的步骤使用不同的 LLM。您可能希望这样做，以便为构思使用更具创造性（即高温度）的模型，为评审和解决使用更严格（即低温度）的模型。

```python
chain = SmartLLMChain(
    ideation_llm=ChatOpenAI(temperature=0.9, model_name="gpt-4"),
    llm=ChatOpenAI(
        temperature=0, model_name="gpt-4"
    ),  # will be used for critique and resolution as no specific llms are given
    prompt=prompt,
    n_ideas=3,
    verbose=True,
)
```
