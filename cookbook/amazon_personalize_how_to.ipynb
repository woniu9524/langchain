### 2.1. Creating a responsive layout

### 2.1. 创建响应式布局

`react-native-web` provides a way to create responsive layouts that adapt to different screen sizes. You can use the `useWindowDimensions` hook to get the current window dimensions and apply styles accordingly.

`react-native-web` 提供了一种创建响应式布局的方法，该布局可以适应不同的屏幕尺寸。您可以使用 `useWindowDimensions` 钩子来获取当前的窗口尺寸并相应地应用样式。

```jsx
import * as React from 'react';
import { Text, View, Button, StyleSheet, useWindowDimensions } from 'react-native';

export default function App() {
  const { width, height } = useWindowDimensions();

  return (
    <View style={[styles.container, { width, height }]}>
      <Text style={styles.text}>Window width: {width}</Text>
      <Text style={styles.text}>Window height: {height}</Text>
      {width > 768 ? (
        <Text style={styles.text}>This is a wide screen</Text>
      ) : (
        <Text style={styles.text}>This is a narrow screen</Text>
      )}
    </View>
  );
}

const styles = StyleSheet.create({
  container: {
    flex: 1,
    justifyContent: 'center',
    alignItems: 'center',
    backgroundColor: '#f5f5f5'
  },
  text: {
    fontSize: 24,
    marginBottom: 20
  }
});
```

In this example, the component displays the current window width and height. It also conditionally renders a text based on whether the screen width is greater than 768 pixels.

在此示例中，组件显示当前的窗口宽度和高度。它还根据屏幕宽度是否大于 768 像素来有条件地渲染文本。

### 2.2. Using platform-specific code

### 2.2. 使用平台特定代码

`react-native-web` allows you to write platform-specific code using the `Platform` module. This is useful when you need to implement different behavior or styles for different platforms.

`react-native-web` 允许您使用 `Platform` 模块编写平台特定的代码。当您需要为不同平台实现不同的行为或样式时，这非常有用。

```jsx
import * as React from 'react';
import { Text, View, Button, StyleSheet, Platform } from 'react-native';

export default function App() {
  const platformText = Platform.select({
    web: 'Hello from the web!',
    ios: 'Hello from iOS!',
    android: 'Hello from Android!'
  });

  return (
    <View style={styles.container}>
      <Text style={styles.text}>{platformText}</Text>
    </View>
  );
}

const styles = StyleSheet.create({
  container: {
    flex: 1,
    justifyContent: 'center',
    alignItems: 'center',
    backgroundColor: '#e0e0e0'
  },
  text: {
    fontSize: 24
  }
});
```

This example uses `Platform.select` to display a different message depending on the platform the app is running on.

此示例使用 `Platform.select` 根据应用程序运行的平台显示不同的消息。

### 2.2 [用例-2] 调用个性化链以总结结果

````python
from langchain.llms.bedrock import Bedrock
from langchain_experimental.recommenders import AmazonPersonalizeChain

bedrock_llm = Bedrock(model_id="anthropic.claude-v2", region_name="us-west-2")

# Create personalize chain
# Use return_direct=True if you do not want summary
chain = AmazonPersonalizeChain.from_llm(
    llm=bedrock_llm, client=client, return_direct=False
)
response = chain({"user_id": "1"})
print(response)
````

### 2.3 [用例-3] 使用您自己的提示调用 Amazon Personalize Chain

````python
from langchain.prompts.prompt import PromptTemplate

RANDOM_PROMPT_QUERY = """
You are a skilled publicist. Write a high-converting marketing email advertising several movies available in a video-on-demand streaming platform next week, 
    given the movie and user information below. Your email will leverage the power of storytelling and persuasive language. 
    The movies to recommend and their information is contained in the <movie> tag. 
    All movies in the <movie> tag must be recommended. Give a summary of the movies and why the human should watch them. 
    Put the email between <email> tags.

    <movie>
    {result} 
    </movie>

    Assistant:
    """

RANDOM_PROMPT = PromptTemplate(input_variables=["result"], template=RANDOM_PROMPT_QUERY)

chain = AmazonPersonalizeChain.from_llm(
    llm=bedrock_llm, client=client, return_direct=False, prompt_template=RANDOM_PROMPT
)
chain.run({"user_id": "1", "item_id": "234"})
````

### 2.4 [用例-4] 在顺序链中调用 Amazon Personalize

````python
from langchain.chains import LLMChain, SequentialChain

RANDOM_PROMPT_QUERY_2 = """
You are a skilled publicist. Write a high-converting marketing email advertising several movies available in a video-on-demand streaming platform next week, 
    given the movie and user information below. Your email will leverage the power of storytelling and persuasive language. 
    You want the email to impress the user, so make it appealing to them.
    The movies to recommend and their information is contained in the <movie> tag. 
    All movies in the <movie> tag must be recommended. Give a summary of the movies and why the human should watch them. 
    Put the email between <email> tags.

    <movie>
    {result}
    </movie>

    Assistant:
    """

RANDOM_PROMPT_2 = PromptTemplate(
    input_variables=["result"], template=RANDOM_PROMPT_QUERY_2
)
personalize_chain_instance = AmazonPersonalizeChain.from_llm(
    llm=bedrock_llm, client=client, return_direct=True
)
random_chain_instance = LLMChain(llm=bedrock_llm, prompt=RANDOM_PROMPT_2)
overall_chain = SequentialChain(
    chains=[personalize_chain_instance, random_chain_instance],
    input_variables=["user_id"],
    verbose=True,
)
overall_chain.run({"user_id": "1", "item_id": "234"})
````
