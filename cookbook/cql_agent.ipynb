### Python 模块

Python 模块是包含 Python 定义和语句的文件。文件名就是模块名加上 `.py` 后缀。模块可以被导入到其他模块或在交互式解释器中使用。

#### 导入模块

你可以使用 `import` 语句来导入模块。

```python
import mymodule
```

导入后，你可以通过模块名来访问其中的函数和变量：

```python
mymodule.my_function()
```

你也可以使用 `from ... import` 语句来导入模块中的特定部分：

```python
from mymodule import my_function
```

导入后，你可以直接使用函数名：

```python
my_function()
```

你还可以使用 `as` 关键字为导入的模块或其部分指定别名：

```python
import mymodule as mm
```

然后使用别名访问：

```python
mm.my_function()
```

#### 模块搜索路径

当导入一个模块时，Python 会按照一定的顺序搜索模块。这个搜索路径包括：

1.  当前目录。
2.  环境变量 `PYTHONPATH` 中列出的目录。
3.  Python 安装的默认库目录。

你可以通过 `sys` 模块来查看当前的搜索路径：

```python
import sys
print(sys.path)
```

#### 创建自己的模块

创建自己的模块非常简单，只需创建一个 `.py` 文件即可。例如，创建一个名为 `greeting.py` 的文件：

```python
# greeting.py
def say_hello(name):
    return f"Hello, {name}!"

def say_goodbye(name):
    return f"Goodbye, {name}!"
```

然后，你可以在另一个 Python 文件中导入并使用它：

```python
# main.py
import greeting

message = greeting.say_hello("Alice")
print(message)
```

#### 命名空间

每个模块都有自己的私有命名空间。这意味着模块中的变量和函数不会与主程序或其他模块中的同名变量和函数发生冲突。

#### `__name__` 变量

每个 Python 模块都有一个内置变量 `__name__`。当模块被直接运行时，`__name__` 的值为 `"__main__"`。当模块被导入到其他模块时，`__name__` 的值为模块的名称。

这允许你编写可以在被导入时执行特定代码，而在直接运行时执行另一段代码的模块。

```python
# my_module.py
def greet():
    print("This is my_module.")

if __name__ == "__main__":
    print("my_module is being run directly.")
    greet()
else:
    print("my_module is being imported.")
```

当你直接运行 `my_module.py` 时，输出将是：

```
my_module is being run directly.
This is my_module.
```

当你从另一个文件导入 `my_module` 时，输出将是：

```
my_module is being imported.
```

#### 包 (Packages)

包是组织 Python 模块的一种方式。一个包就是一个包含多个模块的目录，并且该目录必须包含一个名为 `__init__.py` 的文件（即使该文件为空）。

例如，一个简单的包结构可能如下：

```
my_package/
    __init__.py
    module1.py
    module2.py
```

你可以这样导入包中的模块：

```python
import my_package.module1
```

或者导入特定模块中的函数：

```python
from my_package.module2 import some_function
```

`__init__.py` 文件可以用来初始化包，或者定义包的 `__all__` 变量，该变量指定了当使用 `from my_package import *` 时应该导入哪些模块。

```python
# my_package/__init__.py
from .module1 import function_from_module1
from .module2 import function_from_module2

__all__ = ["function_from_module1", "function_from_module2"]
```

*   **Node.js:** Ensure you have Node.js installed on your system. You can download it from [nodejs.org](https://nodejs.org/).
*   **Database Software:** You will need a database system installed. This guide assumes you are using PostgreSQL, but the principles apply to other relational databases as well.
    *   **PostgreSQL:** Download and install PostgreSQL from [postgresql.org](https://www.postgresql.org/).

*   **Node.js:** 请确保您的系统已安装 Node.js。您可以从 [nodejs.org](https://nodejs.org/) 下载。

### Steps

### 步骤

1.  **Install PostgreSQL (if not already installed):**
    Follow the installation instructions for your operating system from the PostgreSQL website.

2.  **Create a Database:**
    Open your PostgreSQL client (e.g., `psql` or pgAdmin) and execute the following SQL command to create a new database:

    ```sql
    CREATE DATABASE mydatabase;
    ```

    Replace `mydatabase` with your desired database name.

3.  **Create a User (Optional but Recommended):**
    It's good practice to create a dedicated user for your application instead of using the default superuser.

    ```sql
    CREATE USER myuser WITH PASSWORD 'mypassword';
    ```

    Replace `myuser` and `mypassword` with your desired username and password.

4.  **Grant Privileges:**
    Grant the necessary privileges to your user on the newly created database.

    ```sql
    GRANT ALL PRIVILEGES ON DATABASE mydatabase TO myuser;
    ```

    Make sure to replace `mydatabase` and `myuser` accordingly.

5.  **Configure Your Application:**
    In your application's configuration file (e.g., `.env` or `config.js`), update the database connection settings with the credentials you just created:

    ```
    DB_HOST=localhost
    DB_PORT=5432
    DB_USER=myuser
    DB_PASSWORD=mypassword
    DB_NAME=mydatabase
    ```

    Adjust `DB_HOST` and `DB_PORT` if your database is running on a different machine or port.

6.  **Run Migrations (if applicable):**
    If your project uses a migration tool (like Knex.js, Sequelize, or TypeORM), run your database migrations to set up the necessary tables and schemas.

    For example, with Knex.js:

    ```bash
    npx knex migrate:latest
    ```

    Consult your project's documentation for specific migration commands.

### 将数据加载到 Cassandra

本节假定存在一个名为 `df` 的数据框，以下单元格将验证其结构。上面的“下载”部分创建了此对象。

````python
assert df is not None, "Dataframe 'df' must be set"
expected_columns = [
    "ts",
    "device",
    "co",
    "humidity",
    "light",
    "lpg",
    "motion",
    "smoke",
    "temp",
]
assert all(
    [column in df.columns for column in expected_columns]
), "DataFrame does not have the expected columns"
````

创建和加载表：

````python
from datetime import UTC, datetime

from cassandra.query import BatchStatement

# Create sensors table
table_query = """
CREATE TABLE IF NOT EXISTS iot_sensors (
    device text,
    conditions text,
    room text,
    PRIMARY KEY (device)
)
WITH COMMENT = 'Environmental IoT room sensor metadata.';
"""
session.execute(table_query)

pstmt = session.prepare(
    """
INSERT INTO iot_sensors (device, conditions, room)
VALUES (?, ?, ?)
"""
)

devices = [
    ("00:0f:00:70:91:0a", "stable conditions, cooler and more humid", "room 1"),
    ("1c:bf:ce:15:ec:4d", "highly variable temperature and humidity", "room 2"),
    ("b8:27:eb:bf:9d:51", "stable conditions, warmer and dryer", "room 3"),
]

for device, conditions, room in devices:
    session.execute(pstmt, (device, conditions, room))

print("Sensors inserted successfully.")

# Create data table
table_query = """
CREATE TABLE IF NOT EXISTS iot_data (
    day text,
    device text,
    ts timestamp,
    co double,
    humidity double,
    light boolean,
    lpg double,
    motion boolean,
    smoke double,
    temp double,
    PRIMARY KEY ((day, device), ts)
)
WITH COMMENT = 'Data from environmental IoT room sensors. Columns include device identifier, timestamp (ts) of the data collection, carbon monoxide level (co), relative humidity, light presence, LPG concentration, motion detection, smoke concentration, and temperature (temp). Data is partitioned by day and device.';
"""
session.execute(table_query)

pstmt = session.prepare(
    """
INSERT INTO iot_data (day, device, ts, co, humidity, light, lpg, motion, smoke, temp)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
"""
)


def insert_data_batch(name, group):
    batch = BatchStatement()
    day, device = name
    print(f"Inserting batch for day: {day}, device: {device}")

    for _, row in group.iterrows():
        timestamp = datetime.fromtimestamp(row["ts"], UTC)
        batch.add(
            pstmt,
            (
                day,
                row["device"],
                timestamp,
                row["co"],
                row["humidity"],
                row["light"],
                row["lpg"],
                row["motion"],
                row["smoke"],
                row["temp"],
            ),
        )

    session.execute(batch)


# Convert columns to appropriate types
df["light"] = df["light"] == "true"
df["motion"] = df["motion"] == "true"
df["ts"] = df["ts"].astype(float)
df["day"] = df["ts"].apply(
    lambda x: datetime.fromtimestamp(x, UTC).strftime("%Y-%m-%d")
)

grouped_df = df.groupby(["day", "device"])

for name, group in grouped_df:
    insert_data_batch(name, group)

print("Data load complete")
````
